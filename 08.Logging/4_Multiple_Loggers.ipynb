# Multiple Loggers

- When your project grows large (big app, multiple files, modules)
- Using a single root logger is not enough

- make a separete Logger for each Module 
- That way we could differernciate the source of log ( i.e from Module 1 )

---

## synatx

    1st: Create Logger and Set its level
    2nd: Create File Handler, Set its Formating and Add File Handler to Logger

````python
# create logger 

logger1 = logging.getLogger("Module 1")
logger1.setLevel(logging.DEBUG)

# create Handler  ( create Handler and stor it in variable)
logger1_file = logging.FileHandler("Multiloger_Module_1.log")
logger1_console = logging.StreamHandler()

# create Formatter     
formatter = logging.Formatter("%(asctime)s - %(name)s - %(levelname)s - %(message)s ")
# set Formatter to Handler ( add formatting to ur file Handler )
logger1_file.setFormatter(formatter)
logger1_console.setFormatter(formatter)

# Add Handler to loger   ( add those Handler to Logger )
logger1.addHandler(logger1_file)
logger1.addHandler(logger1_console)

# Using 

logger1.debug("This is a Debug msg")
logger1.info("this is an Info msg")
logger1.error("this is an error msg")
logger1.warning("this is WARNING")

# Notice in Out now instead of Root we see Module 1 
````

````python
import logging

formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')

# ================== LOGGER 1: AUTH ==================
logger1 = logging.getLogger("Module 1 ")
logger1.setLevel(logging.DEBUG)

logger1_file = logging.FileHandler("Module_1_Logs.log")
logger1_console = logging.StreamHandler()

logger1_file.setFormatter(formatter)
logger1_console.setFormatter(formatter)

logger1.addHandler(logger1_file)
logger1.addHandler(logger1_console)


# ================== LOGGER 2: DATABASE ==================
logger2 = logging.getLogger("Module 2 ")
logger2.setLevel(logging.DEBUG)

logger2_file = logging.FileHandler("Module_2_logs.log")
logger2_console = logging.StreamHandler()

logger2_file.setFormatter(formatter)
logger2_console.setFormatter(formatter)

logger2.addHandler(logger2_file)
logger2.addHandler(logger2_console)

# ================== LOGGER 3: NETWORK ==================
logger3 = logging.getLogger("Module 3")
logger3.setLevel(logging.INFO)

logger3_file = logging.FileHandler("Module_3_logs.log")
logger3_console = logging.StreamHandler()

logger3_file.setFormatter(formatter)
logger3_console.setFormatter(formatter)

logger3.addHandler(logger3_file)
logger3.addHandler(logger3_console)


# -------- TEST LOGS --------

logger1.debug("DEbug msg")
logger1.error("errro msg")
logger2.warning("Warning")
logger2.critical("Shit got srs")
logger2.info("just minding my own shit")
logger3.warning("Baaz AA ja")
logger3.critical("YO whats happenin'")
logger1.critical("i'm outta here Sayo Nara")
````

````python
import logging

# -------- FORMATTER --------
formatter = logging.Formatter('%(levelname)s - %(name)s - %(message)s')



# ================== LOGGER 1: AUTH ==================
auth_logger = logging.getLogger("Auth")
auth_logger.setLevel(logging.DEBUG)

#create handler and store it in a variable
auth_file = logging.FileHandler("auth_system_logs.log")
auth_console = logging.StreamHandler()

# add formatter to above files ( so that they would look like this )
auth_file.setFormatter(formatter)
auth_console.setFormatter(formatter)

## now add these handler to logger 
auth_logger.addHandler(auth_file)
auth_logger.addHandler(auth_console)



# ================== LOGGER 2: DATABASE ==================
db_logger = logging.getLogger("DataBAse")
db_logger.setLevel(logging.DEBUG)

# create file Handler set its format and add these handler to logger
db_file = logging.FileHandler("Database_system_logs.log")
db_console = logging.StreamHandler()

db_file.setFormatter(formatter)
db_console.setFormatter(formatter)

db_logger.addHandler(db_file)
db_logger.addHandler(db_console)

# ================== LOGGER 3: NETWORK ==================
net_logger = logging.getLogger("Network")
net_logger.setLevel(logging.WARNING)

# create File Handler set its format and add it to logger
net_file = logging.FileHandler("Network_system_logs.log")
net_console = logging.StreamHandler()

net_file.setFormatter(formatter)
net_console.setFormatter(formatter)

net_logger.addHandler(net_file)
net_logger.addHandler(net_console)

# -------- USING LOGGERS --------

# AUTH
auth_logger.debug("This will NOT show (level INFO)")
auth_logger.info("User logged in")
auth_logger.warning("Suspicious login attempt")

# DATABASE
db_logger.debug("Query executed")
db_logger.info("Data inserted")
db_logger.error("Database connection failed")

# NETWORK
net_logger.info("This will NOT show (level WARNING)")
net_logger.warning("Slow response detected")
net_logger.error("Network timeout")
````
