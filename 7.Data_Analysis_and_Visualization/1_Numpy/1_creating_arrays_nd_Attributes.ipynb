## Using arange() ##

* automatically create an array from starting range to ending range

### Reshape(rows, cols) ###
- change given array to some other shape
- often used with arange()
-  imp rule: row and cols product must be equal to nmbr of elements present

        np.arange(start,end).reshape(row,col)   
        
        -->> np.arange(1,11).reshape(2,5) -->> (make it in 2 rows and 5 cols ) also 2*5 = 10 == nmbr of ele

```python
#dtype -->> we can make prefered data type array using dtype

a = np.array([1,2,3], dtype = float)
print(a) 

b = np.array([0, 1,2,4], dtype = bool)
print(b) # as 0 == False , other nmbrs == True

c = np.array([1,2,3], dtype= complex)
print(c)
```

# Changing Datatype of nparray #
- astype

                array.astype(datatype)
        
* array.astype(np.int32)

```python
array1 = np.arange(1,11)
array2 = np.arange(0,12, dtype=float).reshape(3,4)
array3 = np.arange(8).reshape(2,2,2)

print(array1)

print()

print(array2)

print()

print(array3)
```

```python
# ndim

print(array1.ndim)
print(array2.ndim)
print(array3.ndim)
```

```python
# shape

print(array1.shape)
print(array2.shape)
print(array3.shape)
```

```python
# size
print(array1.size)
print(array2.size)
print(array3.size)
```

```python
# itemsize 

print(array1.itemsize)
print(array2.itemsize) # as float and int occupy same space
print(array3.itemsize)
```

```python
#dtype
print(array1.dtype)
print(array2.dtype)
```

```python
#astype

print(array1.dtype)
# now change

array1.astype(np.int32)

```
