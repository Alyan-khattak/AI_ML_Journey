````python
# Circular Referencing 

import gc

a = [] # # Variable a  →  Object []
b = []

a.append(b)
b.append(a)

del a # REMOVE the variable name "a": del removes the REFERENCE (name), not the object itself
del b

"""
a points to list1
b points to list2

list1 contains reference to list2
list2 contains reference to list1

del deleted the refrene not object (list)
Variable "a" removed
Variable "b" removed

but still

list1 → list2
list2 → list1

list1 has reference from list2
list2 has reference from list1

########################################
a → [   ]   (list object)

a.append(b)

means:

[   ].append(b)

#########3333333333333333333333333333333############

Ref count ≠ 0 ❌

Python only deletes when ref count = 0


#################################################3
a = person A
b = person B

A holds B
B holds A

Now you leave the room (del a, del b)

But A and B are still holding each other 🤝


"""

collected = gc.collect()
print("Objects collected:", collected)

# GC removed 2 objects (a and b)
````

````python
# disable gc 
gc.disable()

gc.disable()

a = [] 
b = []

a.append(b)
b.append(a)

del a
del b

print("GC Disabled")

collected = gc.collect()
print("Collected:", collected)
````

````python
class Myobjct:
    def __init__(self, name):
        self.name = name
        print(f"Object {self.name} created ")

    def __del__ (self):
        print(f"Object  {self.name} deleted")

# circular reference

obj1 = Myobjct("object1")
obj2 = Myobjct("object2")

obj1.ref = obj2
obj2.ref = obj1

del obj1
del obj2 # now both objects will be unusable but still not delted cuz of circular ref

# we didnt get the deleted msg cuz object didnt got deleted
        
````

````python
class Myobjct:
    def __init__(self, name):
        self.name = name
        print(f"Object {self.name} created ")

    def __del__ (self):
        print(f"Object  {self.name} deleted")

# circular reference

obj1 = Myobjct("object1")
obj2 = Myobjct("object2")

obj1.ref = obj2
obj2.ref = obj1

del obj1
del obj2 # now both objects will be unusable but still not delted cuz of circular ref


gc.collect()

# now its show delted msg why 4 times becuz one ref for each obj from above was also not cleaned
````

````python
# Real-Life Example: Users & Friend Circles

import gc

class User:
    def __init__(self, name):
        self.name = name
        self.friends = [] # list of frnd object
        print(f"User {self.name} created")

    def add_frnd(self, frnd):
        self.friends.append(frnd)

    def __del__(self):
        print(f"user {self.name} Deleted")

# --- Create Users ---
alyan= User("Alyan")
sana = User("sana")
adil = User("adil")


# --- Create Friendships ---

alyan.add_frnd(sana)
sana.add_frnd(adil)
adil.add_frnd(alyan)

# --- Check Reference Counts ---
import sys

print("Ref Count of Alyan : ", sys.getrefcount(alyan))
print("Ref Count of Sana : ", sys.getrefcount(sana))
print("Ref Count of Adil : ", sys.getrefcount(adil))

         

del adil
del alyan
del sana


print("Deleted main references. Objects still exist due to circular refs?")

# --- Force Garbage Collection ---
gc.collect()
print("Garbage collection done")
````
