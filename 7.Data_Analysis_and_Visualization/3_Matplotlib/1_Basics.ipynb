# Outliers in Data Visualization (Matplotlib)

## 🔹 What is an Outlier?

An **outlier** is a data point that is significantly different from the rest of the data.

- It can be **much larger** or **much smaller** than other values.
- It does **not follow the general pattern** of the dataset.

---

## 🔹 Example

Values:
1, 3, 6, 7, 99999999999999

- The last value (`99999999999999`) is an **extreme outlier**.
- It is far away from the rest of the values.

---

## 🔹 Effect on Graphs (Matplotlib)

When plotting:

- The outlier **stretches the axis scale**
- Smaller values (1, 3, 6, 7) get compressed
- The graph may look like a **flat line** for normal values

👉 This happens because matplotlib adjusts the axis to include all values.

---

## 🔹 Visualization Problem

Without outlier:
- Clear variation between values

With outlier:
- Normal values become hard to distinguish
- Important patterns may be hidden

---


## 🔹 Common Solutions

1. **Remove Outliers**
   - If they are errors or irrelevant

2. **Use Log Scale**
   - Helps reduce the effect of large values

3. **Zoom / Limit Axis**
   ```python
   plt.ylim(0, 10)


---

### Limit Axis:

## 🔹 What is Axis Limiting?

Axis limiting means **manually controlling the visible range** of a graph.

Instead of letting matplotlib automatically decide the scale, you **set boundaries** for x-axis or y-axis.

---

## 🔹 Why Do We Use It?

When there are **outliers**, matplotlib stretches the axis:

Example:
1, 3, 6, 7, 99999999999999

- The large value forces the y-axis to go very high
- Smaller values look like a flat line

👉 Axis limiting helps us **focus on important data** by ignoring extreme values visually.

---

## 🔹 Syntax

### Limit Y-axis and X-axis
```python
plt.ylim(min_value, max_value)
plt.xlim(min_value, max_value)


````python
# simple Graph 
price = [48000, 54000, 49000, 45000, 42000, 99999999] #last value is an outlier
years = [2012, 2013, 2016, 2017,2020, 2021]

plt.plot(years, price , color = "red", marker="x", linewidth=2, markersize=20)
plt.xlabel("Years")
plt.ylabel("Price")
plt.title("Yearly Price of Phones")
plt.grid() # grid will show on graph
plt.show()
## The previos points looks like straight line 
````

````python
# above Grapgh using limits on X-axis
# simple Graph 
price = [48000, 54000, 49000, 45000, 42000, 999999999]
years = [2012, 2013, 2016, 2017,2020, 2021]

plt.plot(years, price , color = "red", marker="x", linewidth=2, markersize=20)
plt.xlim(2012,2020) # as Outlier came in 2021 , so we filter it 
plt.ylim(42000,54000) # for max_value we select one less than outlier
plt.xlabel("Years")
plt.ylabel("Price")
plt.title("Yearly Price of Phones")
plt.grid() # grid will show on graph
plt.show()
````

# Scatter Plot in Matplotlib

## 🔹 What is a Scatter Plot?

A **scatter plot** is a graph used to **visualize the relationship between two variables**.

- Each point represents one observation `(x, y)`.
- A dot where x,y intersects 
- ( in line Graph the marker's dot is scatter graph)
- Useful to see:
  - Correlation (positive, negative, none)
  - Clusters or groups
  - Outliers

---

## 🔹 Syntax

Basic syntax:

```python
plt.scatter(x, y, s=size, c=color, marker=marker_style, alpha=transparency)
```
---

## How to Take Insights

### Correlation

- Upwards trend → positive correlation

- Downwards trend → negative correlation

- No trend → no correlation

### Clusters

- Points grouped together → similar behavior

- Separate clusters → different categories

### Outliers

- Points far away from others → unusual observations

### Density

- Many points close → high density

- Few points → low density


## Hidden Parameter:

    scatter(x,y , s=size)

- s = Size of each point , how big a point appear on screen


````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 5]

plt.scatter(x, y)
plt.title("Simple Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
````

```python

# 📊 Side-by-Side Bar Chart (Multiple Bars)

## 🔹 Problem

When plotting multiple bars for the same category:

- By default, `plt.bar()` **stacks bars on top of each other**.
- We want **bars to appear side by side** for easier comparison.

---

## 🔹 Solution: Shift Bars Using Width

We use the **bar width** and **shift positions**:

or that we need to do some calculations 

```python
plt.bar(x - width, y1, width=width)  # 1st bar
plt.bar(x, y2, width=width)          # 2nd bar (centered)
plt.bar(x + width, y3, width=width)  # 3rd bar

```

---

## what we did:

- 1st we set a widh of each bar 
        then

- from 1st bar we subtrack that widh ( so it would pull back a little)
- for 2nd bar we didnt do anything , it says in its place
- from 3rd bar we added with width ( so that it starts from 2nd bar)

---

- suppose or bar would plot on **ONE** with widh 0.2
- so one bar would take width from **1->1.2** right 

- what we did we sbtract that from 1st bar now 1st bar would start from **0.98 ->1** 
- 2nd bar would start from **1->1.2** as its normall
- 3rd bar would start from **1.2 -> 1.4** as normally its starts from **ONE** but we have added witdth with 3rd graph **0.2** so it would start from **1.2**

that the whole mind map


````python
# Multiple Graphs
plt.bar(np.arange(df.shape[0]) - 0.2 , df["2015"], width = 0.2)
plt.bar(np.arange(df.shape[0]), df["2016"], width= 0.2)
plt.bar(np.arange(df.shape[0]) +0.2, df["2017"], width = 0.2)
plt.xlabel("Batsman")
plt.ylabel("Runs")
plt.title("Season Wise Runs")

plt.show()
````

# 📊 Stacked Bar Charts in Matplotlib

## 🔹 What is a Stacked Bar Chart?

A **stacked bar chart** is used when you want to **show how different components contribute to a total** for each category.  

- Each bar is **divided into segments** (stacked on top of each other).  
- Useful for comparing **parts of a whole** across categories.  

---

## 🔹 Key Keyword: `bottom`

- `bottom` specifies **where the bar segment starts on the y-axis**.  
- Default = 0 (starts at the bottom).  
- For stacking:
  - Each new bar segment starts at the **sum of all previous segments** for that category.  

---

## 🔹 Example: Stacked Bar Chart

```python id="stacked_example"
import matplotlib.pyplot as plt

# Example DataFrame
import pandas as pd
df = pd.DataFrame({
    "batsman": ["Player A", "Player B", "Player C"],
    "2015": [5, 7, 6],
    "2016": [3, 4, 5],
    "2017": [6, 8, 7]
})

# Plot stacked bars
plt.bar(df["batsman"], df["2017"], label="2017")                          # Bottom = 0 by default
plt.bar(df["batsman"], df["2016"], bottom=df["2017"], label="2016")       # Start on top of 2017
plt.bar(df["batsman"], df["2015"], bottom=(df["2016"] + df["2017"]), label="2015")  # Start on top of 2016+2017

plt.ylabel("Runs")
plt.title("Runs by Batsman (Stacked by Year)")
plt.legend()
plt.show()

````python
plt.bar(df["batsman"], df["2017"], label = "2017")
plt.bar(df["batsman"], df["2016"], bottom= df["2017"], label="2016")
plt.bar(df["batsman"], df["2015"], bottom= ( df["2016"] + df["2017"] ), label = '2015' )

plt.legend()
plt.show()
````

# 📊 Histogram in Matplotlib

## 🔹 What is a Histogram?

A **histogram** is used to visualize the **distribution of continuous numerical data**.

- It groups data into **bins (ranges)**  
- Shows **how many values fall into each range**  

👉 Unlike bar charts:
- Bar chart → categorical data  
- Histogram → continuous data  

---

### Main Purpose: Frequency Count in a given Range

## 🔹 Example Intuition

Data:
[1, 2, 2, 3, 3, 3, 4, 5]

Histogram groups into bins like:
- 1–2 → count = 3  
- 3–4 → count = 4  
- 5–6 → count = 1  

 Instead of individual values, we see **frequency distribution**

---

## 🔹 Why Use Histogram?

- Understand **data distribution**
- Detect **outliers**
- Identify **skewness (left/right)**
- See **spread (variance)**
- Useful in **EDA (Exploratory Data Analysis)**

---

## 🔹 Use in AI / Data Science

- Check if data is **normally distributed**
- Understand **feature distribution**
- Detect **data imbalance**
- Preprocessing decisions (scaling, transformation)

---

## 🔹 Basic Syntax

```python
plt.hist(data)



plt.hist(
    data,           # input data (list, array, Series)
    bins=None,      # number of bins or specific bin ranges
    range=None,     # lower and upper range
    density=False,  # normalize to probability density
    weights=None,   # weights for each data point
    cumulative=False, # cumulative histogram
    bottom=None,    # starting point of bars
    histtype='bar', # type: 'bar', 'step', 'stepfilled'
    align='mid',    # alignment of bars
    orientation='vertical', # 'horizontal' also possible
    rwidth=None,    # relative width of bars
    log=False,      # log scale
    color=None,     # bar color
    edgecolor=None, # border color
    alpha=None,     # transparency
    label=None      # legend label
)

````python
data = [1,2,2,3,3,3,4,5]

plt.hist(data)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("simple HistoGram")
plt.grid()
plt.show()
````

# 🔹 Using Custom Bin Ranges in Histogram

# 📊 Log Scale in Data Visualization

## 🔹 What is Log Scale?

A **log scale** is a way of plotting data where values are transformed using logarithms.

Instead of plotting actual values, we plot:

log(value)

---

## 🔹 Linear vs Log Scale

### Linear Scale
- Equal spacing = equal difference
- Example:
  1 → 2 → 3 → 4

### Log Scale
- Equal spacing = equal ratio (multiplication)
- Example:
  1 → 10 → 100 → 1000

---

## 🔹 Example

Data:
[1, 10, 100, 1000]

Log transformation:

log(1) = 0  
log(10) = 1  
log(100) = 2  
log(1000) = 3  

👉 Becomes evenly spaced

---

## 🔹 Why Use Log Scale?

- Handles **large ranges of data**
- Makes **small values visible**
- Reduces impact of **outliers**
- Helps analyze **exponential growth**

---

## 🔹 Intuition

- Linear scale → addition (+)
- Log scale → multiplication (×)

👉 Log scale focuses on **relative change**, not absolute change

---

## 🔹 Key Behavior

- Large values are **compressed**
- Small values are **expanded**
- Patterns become easier to see

---

## 🔹 When to Use

✔ Data ranges from very small to very large  
✔ Data is skewed  
✔ Exponential growth patterns  

❌ Avoid when:
- Data has zero or negative values (log undefined)

---

## 🔹 Key Insight

> Log scale changes how we see the data — it reveals patterns hidden in extreme ranges.

---

## Syntax:

    plt.hist(data, log = True)   

- log = True : Aplly log scale on Y-axis

    > If we want log Scale on both x, and y axis then use plt.xscale('log'), plt.yscal('log') 


        plt.xscale('log')
        plt.yscale('log')

````python
# Without Log scale


data = [1]*50 + [2]*40 + [3]*30 + [1000]*2 + [5000]*1

plt.hist(data, bins=5, edgecolor='black')
plt.title("Histogram (Normal Scale)")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
````

# 🥧 Pie Chart in Matplotlib

## 🔹 What is a Pie Chart?

A **pie chart** is used to represent **data as proportions of a whole**.

- Each slice represents a **percentage contribution**
- Total = **100%**
- Good for showing **composition**

---

## 🔹 When to Use Pie Chart?

✔ When you want to show:
- Proportions (e.g., market share, class distribution)
- Percentage contribution
- Parts of a whole

---

## 🔹 Use in AI / Data Science

- Visualize **class distribution**
- Show **feature contribution**
- Compare **model prediction proportions**
- Understand **imbalanced datasets**

---

## 🔹 Basic Syntax

```python
plt.pie(Data)


plt.pie(
    x,              # values
    explode=None,   # Pull slices Outward 
    labels=None,    # category labels
    colors=None,    # slice colors
    autopct=None,   # show percentage  -->>  imp   autopct='%1.1f%%'
    pctdistance=0.6,# position of percentage text
    shadow=False,   # add shadow
    startangle=0,   # rotation angle
    radius=1,       # size of pie
    counterclock=True, # direction
    wedgeprops=None,# slice properties
    textprops=None, # text styling
    center=(0,0),   # center position
    frame=False     # draw frame
)

# To Show Percentage 

autopct='%1.1f%%'

````python
values = [40, 30, 20, 10]
labels = ['A', 'B', 'C', 'D']

plt.pie(values, labels=labels)
plt.legend()
plt.show()
````

````python
# SHow Perentage 
values = [40, 30, 20, 10]
labels = ['A', 'B', 'C', 'D']

plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart with Percentages")
plt.legend()
plt.show()
````

# 📊 Matplotlib: plt.figure()

## 🔹 What is `plt.figure()`?

`plt.figure()` in Matplotlib is used to **create a new figure object**.

- A **figure** is like a **canvas** or **container** for all your plots.
- You can have **multiple figures** in a single notebook or script.
- Each figure can contain **one or more axes** (subplots).

---

## 🔹 Why use `plt.figure()`?

1. **Control figure size**: Make plots larger/smaller
2. **Multiple figures**: Plot several graphs independently
3. **High-quality saving**: Save individual figures
4. **Advanced customization**: Set DPI, background color, etc.

---

## 🔹 Basic Syntax

```python
plt.figure(
    num=None,        # Figure number or name
    figsize=None,    # Width, height in inches (tuple)
    dpi=None,        # Resolution (dots per inch)
    facecolor=None,  # Background color
    edgecolor=None,  # Border color
    tight_layout=False,  # Adjust layout automatically
    constrained_layout=False  # Use tight layout automatically
)

# important 
figsize(5,6) # figure size in inches

```
---
| Parameter            | Purpose                                 |
| -------------------- | --------------------------------------- |
| `num`                | Name/number of the figure               |
| `figsize`            | Tuple `(width, height)` in inches       |
| `dpi`                | Resolution of figure (higher = sharper) |
| `facecolor`          | Background color                        |
| `edgecolor`          | Border color                            |
| `tight_layout`       | Adjust spacing automatically            |
| `constrained_layout` | More advanced layout adjustment         |


---
