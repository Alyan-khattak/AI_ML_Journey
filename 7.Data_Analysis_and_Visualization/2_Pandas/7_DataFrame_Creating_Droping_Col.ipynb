# Creating Column
- two types

    - Completely new col in Dataset
    - existing columns → new useful column

## Completely New Col

    df["col_name"] = values

## New Col From Existing Cols
- Here we make new Col from existing Col by using Operations, Consitions etc

- exm:

        def["new_col"] = df["Old_col"] + df["Old_col2"]
        df["new_col"] = df["old_column"] > 7


# Droping Col and Row
## Col
    df.drop("column_name", axis=1)

- axis = 1 means Col
- azis = 0 means Row

## row

    df.drop(2)

- whole row at index 2 will be dropped
- After Deleting rows the indexing needs to be reset so we use

        df.reset_index(drop=True)

```python
import pandas as pd
```

```python
movies = pd.read_csv("/home/aizen/AI_ML/7.Data_Analysis_and_Visualization/2_Pandas/movies.csv")
movies.head(2)
```

```python
# Completely New Col

movies["Country"] = "India"
movies.head(2)
movies["Country"].head(2)   # new Col 


```

```python
# from Exiting Col

movies["hit_movies"] = movies["imdb_rating"] > 8
movies.head(2) # now Col has been added
```

```python
# long Movie Col
movies["runtime"] = pd.to_numeric(movies["runtime"], errors="coerce")
movies["Long_movie"] = movies["runtime"] > 200
movies.head(2)
```

```python
# genere Count Col
movies["genres_Count"] = movies["genres"].str.split("|").str.len()
movies.head(2)
```

```python
# Data Frame 

data = {
    "name": ["Alyan", "Adil", "Asad", "Aroosa", "Alien"],
    "age": [22, 21, 23, 20, 24],
    "city": ["Lahore", "Karachi", "Islamabad", "Lahore", "Karachi"],
    "salary": [50000, 60000, 55000, 52000, 65000]
}

df = pd.DataFrame(data)
df
```

```python
# insertng a Column
df["bonus"] = df["salary"] * 0.10
df["name_length"] = df["name"].str.len()
df
```

```python
df.drop("name_length", axis=1, inplace=True)
df # we drooped the name_length Cols
```

```python
df.drop(4, inplace=True) # 4th row from all cols will be delete
df.drop(1, inplace=True)
df # we see the indexing is wrong .. so to reset index we use df.reset_index(drop=True)
```

```python
df.reset_index(drop=True)
```

```python
data = {
    "student_id": [101,102,103,104,105,106,107,108],
    "name": ["Ali","Sara","Ahmed","Zara","Usman","Hina","Bilal","Ayesha"],
    "age": [21,22,20,23,24,21,22,20],
    "city": ["Lahore","Karachi","Islamabad","Lahore","Karachi","Lahore","Islamabad","Karachi"],
    "math_score": [78,88,92,65,70,85,90,60],
    "english_score": [80,75,85,60,72,88,91,65],
    "study_hours": [3,5,6,2,3,4,5,2]
}

df1 = pd.DataFrame(data)
df1
```

```python
df1["total_score"] = df1["english_score"] + df1["math_score"]
df1["avg_score"] = df1["total_score"] / 2
df1["is_toper"] = df1["avg_score"] > 90
df1
```

```python
# A col using Cotigory 
# age >= 22 → "Senior"
# age < 22 → "Junior"

# use np.where(condition, if true, if false)
import numpy as np
df1["age_category"] = np.where(df1["age"] >= 22, "senoir",  "Junior")
df1
```

```python
# Droping Col
df1.drop("student_id", axis=1)
df1.drop("name", axis=1)
```

```python
# Row Filtering
# keep only students with Scores > 70

df1 = df1[df1["math_score"] > 70]
df1
```

```python
df1[df1["avg_score"] >= 75]
df1 = df1.reset_index(drop=True)
df1
```

```python
df1 = df1.sort_values("avg_score", ascending=False)
df1 = df1.reset_index(drop=True)
df1
```
