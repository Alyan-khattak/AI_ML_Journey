```python
"""project/
    data/
        file1.txt
        file2.txt
""" # making the above structures

base = "project"
data_folder = os.path.join(base, "Data" )  #base/data

if not os.path.exists(base):  # create base dire
    os.mkdir(base)

if not os.path.exists(data_folder): # create data dir in base 
    os.mkdir(data_folder)

# create files
for i in range(1,3):
    file_path = os.path.join(data_folder, f"file{i}.txt")
    with open(file_path, "w") as file:
        file.write("hello world")

# cout files
total_files = 0
for root, dir, files in os.walk(base):
    total_files += len(files)

print("total files : ",total_files)
    
```

```python
## Find largest File in directory 

folder = "project"

largest_file = None
largest_size = 0

for root, dirs, files in os.walk(folder):
    for file in files:
        filee_path = os.path.join(root, file)
        size = os.path.getsize(filee_path)

        if size > largest_size:
            largest_size = size
            largest_file = filee_path

print("Largest file:", largest_file)
print("Size:", largest_size, "bytes")
```

```python
# find files with same names in diff folders

folder = "project"
seen = {}
duplicates = []

for root, dirs, files in os.walk(folder):
    for file in files:
        if file in seen:
            duplicates.append(os.path.join(root, file))
        else:
            seen[file] = os.path.join(root, file)


print("Duplicate files:")
for dup in duplicates:
    print(dup)
```

```python
# Create this structure
""" 
dataset/
    cats/
        img1.jpg
        img2.jpg
    dogs/
        img1.jpg
    birds/

"""

# counts total classes (subfolders)
# Counts total files
# Prints file count per class
# Creates a report file called report.txt inside dataset folder

import os

# Base dataset folder
dataset_path = "dataset"

# -----------------------------
# CREATE FOLDER STRUCTURE
# -----------------------------

# Create main dataset folder if it doesn't exist
if not os.path.exists(dataset_path):
    os.mkdir(dataset_path)

# Define class folders
classes = ["cats", "dogs", "birds"]

for cls in classes:
    # Build full path for each class folder
    # Example: dataset/cats
    class_path = os.path.join(dataset_path, cls)

    # Create class folder if not exists
    if not os.path.exists(class_path):
        os.mkdir(class_path)

# Create dummy image files
# dataset/cats/img1.jpg
# dataset/cats/img2.jpg
# dataset/dogs/img1.jpg

files_to_create = {
    "cats": ["img1.jpg", "img2.jpg"],
    "dogs": ["img1.jpg"],
    "birds": []
}

for cls, filenames in files_to_create.items():
    class_path = os.path.join(dataset_path, cls)

    for filename in filenames:
        file_path = os.path.join(class_path, filename)

        # Create empty file
        with open(file_path, "w") as f:
            f.write("dummy image content")

# ---------------------------------
# COUNT CLASSES AND FILES
# ---------------------------------

total_classes = 0
total_files = 0

# Path for report file
report_path = os.path.join(dataset_path, "report.txt")

with open(report_path, "w") as report:

    # Loop through items inside dataset folder
    for item in os.listdir(dataset_path):

        item_path = os.path.join(dataset_path, item)

        # Check if item is a folder (class)
        if os.path.isdir(item_path):

            total_classes += 1
            file_count = 0

            # Count files inside class folder
            for file in os.listdir(item_path):
                file_path = os.path.join(item_path, file)

                if os.path.isfile(file_path):
                    file_count += 1

            total_files += file_count

            print(f"{item}: {file_count} files")
            report.write(f"{item}: {file_count} files\n")

    report.write(f"\nTotal Classes: {total_classes}\n")
    report.write(f"Total Files: {total_files}\n")

print("\nTotal Classes:", total_classes)
print("Total Files:", total_files)
print("Report created at:", report_path)
```
