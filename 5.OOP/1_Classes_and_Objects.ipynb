```python
class Student:
    
    # CLass Attribute: Shared by all the Object of Class (just like Static in Java)
    uni = "COMSATS WAH"   # Class also acces them using Class Name i.e print-->> Student.uni 

    #parameterized Constructor 
    # Instance Variables: Belong to each objct separatly
    def __init__(self, name, age, marks): 
        self.name = name #self is like "This" keyword we Used in Java
        self.age = age
        self.marks = marks


    #Method 
    def avg(self):
        sum = 0

        for i in self.marks:
            sum += i
        print(f"{self.name} Avg marks are : {sum/3}")


    #Static Method
    @staticmethod
    def hello():
        print("Hello")



s1 = Student("Alyan", 21 , [99,90,39])  #objct 1
print(s1.name)
print(s1.age)
print(s1.marks)
print(s1.uni)  #Accesing Class variable using Objct
print(Student.uni) # Accessing Class Variable usig Class
print()


s2 = Student("Aroosa",20,[98,100,100]) # Objct 2
print(s2.name)
print(s2.age)
print(s2.uni) 

## using method 
print("\n Using Methods \n ")
s1.avg()
s2.avg()

# static Method 
s1.hello()


```

```python
# ATM Machine Working Using OOP

class ATM:

    # default Constructor to initaialize Values
    def __init__(self):
        self.owner = None
        self.balance = 0
        self.pin = None
        self.is_blocked = False

    
    ## Logic METHODS ##

    def create_account(self):
        print("\n--- Create New Account ---")
        self.owner = input("Enter your name: ")
        self.pin = input("Create a 4-digit PIN: ")
        self.balance = float(input("Enter initial balance: "))
        print("\nAccount created successfully!")

    ## Login ##
    def login(self):
        if self.pin is None:
            print("\nNO Acc Found. Plz Create One First")
            return False
    
        if self.is_blocked:
            print("\n Account is Blocked due to 3 wrong PIN Attempts ")
            return False  
        
        attempts = 3
        while attempts > 0:
            entered_pin = input("Enter ur PIN : ")
            if entered_pin == self.pin:
                print("Login Successfully !")
                return True
            
            else: 
                attempts -= 1
                print(f"Wrong PIN Attempts Left {attempts}")


        self.is_blocked = True
        print("\nAccount blocked due to too many wrong attempts.")
        return False
    

    ## Check Balance
    def check_balance(self):
        print(f"\n Curret Balance : {self.balance}")

    ## Deposit

    def deposit(self):
        amount = float(input("enter amout to deposit : "))
        if amount > 0:
            self.balance += amount
            print("Deposit Successfully.")
        else:
            print("Invalid Amount!")

    ## Withdraw
    def withdraw(self):
        amount = float(input("Enter amount to Withdraw: "))
        if amount > self.balance:
            print("Insufficient Funds ")
        elif amount <= 0:
            print("Invalid Amount.")
        else:
            self.balance -= amount
            print("withdrawl Successfull ")


    
    # Menu Method
    def main_menu(self):
        while True:
            print("1. Create an Acc ")
            print("2. Login")
            print("3. Exit")

            choice = int(input("Enter CHoice "))

            if choice == 1:
                if self.pin is None:
                    self.create_account()
                else:
                    print("Acc alrdy Exists ")
            elif choice == 2:
                if self.login():
                    self.user_menu()
            elif choice == 3:
                print("Thank You for Using ATM ")
                break
            else:
                print("Invalid Choice ")

    
    #User Menu Logged in Users

    def user_menu(self):
        while True:
            print("\n--- ATM MENU ---")
            print("1. Check Balance")
            print("2. Deposit")
            print("3. Withdraw")
            print("4. Logout")

            choice = input("Enter choice: ")

            if choice == "1":
                self.check_balance()

            elif choice == "2":
                self.deposit()

            elif choice == "3":
                self.withdraw()

            elif choice == "4":
                print("Logged out successfully.")
                break

            else:
                print("Invalid choice.")


## Create Objects

atm1 = ATM()
atm1.main_menu()

```
