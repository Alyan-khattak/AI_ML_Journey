# Imbalanced Datasets: Complete Notes for Beginners

## What is an Imbalanced Dataset?

**One sentence definition:** When one category in your data has far more examples than another category, causing your model to become "lazy" and ignore the smaller category.

**Simple analogy:** Imagine training a security robot to detect bombs vs lunchboxes with 95 lunchboxes and only 5 bombs. The robot learns to say "lunchbox" every time and is "right" 95% of the time, but completely fails at detecting bombs.

> Model is lazy → it learns what it sees most

| A      | No     |
| B      | No     |
| C      | No     |
| D      | No     |
| E      | Yes    |

👉 Here:

No = 4
Yes = 1

This is imbalanced.


## The Core Problem

| If your dataset is... | The model becomes... | Real-world consequence |
|----------------------|----------------------|------------------------|
| Balanced (50/50) | Fair and smart | Detects everything well |
| Imbalanced (95/5) | Lazy and biased | Ignores rare but important cases |

**Why it's bad:** The model optimizes for overall accuracy, not for catching rare events. But rare events are often the MOST important (fraud, disease, threats).

## Techniques to Solve Imbalanced Datasets

### 1. Get More Data (The Obvious One)
- Collect more examples of the rare class
- **Limitation:** Sometimes impossible (rare diseases, rare fraud)

### 2. Oversampling (Duplicate the small group)
- Copy rare examples multiple times
- Example: Copy 5 bombs → 100 bombs
- **Risk:** Model might memorize exact copies

### 3. Undersampling (Shrink the big group)
- Randomly remove examples from majority class
- Example: Reduce 95 lunchboxes → 5 lunchboxes
- **Risk:** Lose useful information

### 4. SMOTE (Smarter Oversampling)
- Create *slightly new* synthetic examples of rare class
- Mix features of existing rare examples to create realistic variations
- **Better than simple duplication**

### 5. Use Different Performance Metrics
- **Don't trust accuracy** with imbalanced data
- Use instead:
  - **Recall:** "Out of all real bombs, how many did you catch?"
  - **Precision:** "When you said 'bomb', how often were you right?"
  - **F1-Score:** Harmonic mean of both

### 6. Class Weights (Penalize the model)
- Tell model: "Missing a bomb costs 100x more than missing a lunchbox"
- Forces model to care about rare class
- Built into most ML algorithms (parameter like `class_weight='balanced'`)

# Undersampling: Types, Advantages & Disadvantages (Brief)

## Types of Undersampling

| Type | How It Works | Best For |
|------|--------------|----------|
| **Random Undersampling** | Randomly removes majority samples | Quick baseline, huge datasets |
| **NearMiss-1** | Keeps majority samples closest to minority | When 
| **Tomek Links** | Removes ambiguous samples on class boundary | Cleaning noisy boundaries |
| **Cluster Centroids** | Replaces majority with cluster centers | When you want representative samples, not random |
| **Edited Nearest Neighbors (ENN)** | Removes majority samples misclassified by neighbors | Removing noisy/overlapping samples |

---

## Advantages ✅

| Advantage | Why It Matters |
|-----------|----------------|
| **Faster training** | Less data = quicker computation |
| **Lower memory usage** | Smaller dataset fits in RAM |
| **No overfitting risk** | Not duplicating data like oversampling |
| **Simple to implement** | Just random selection |
| **Reduces storage costs** | Less data to save |
| **Can remove noisy samples** | Some methods (ENN, Tomek) clean data |

---

## Disadvantages ❌

| Disadvantage | Why It Hurts |
|--------------|--------------|
| **Information loss** | Throws away potentially useful data |
| **Missed patterns** | Removed samples might contain important signals |
| **Randomness issues** | Different random seeds give different results |
| **Not for small datasets** | Can't afford to lose any data |
| **Biased representation** | May not represent original distribution |
| **Worse for very rare classes** | Minority < 100 samples → too much loss |

---

## Quick Comparison: When to Use

| Situation | Use Undersampling? |
|-----------|-------------------|
| Huge dataset (millions) | ✅ YES |
| Minority class > 1000 samples | ✅ YES |
| Training is too slow | ✅ YES |
| Small dataset (< 10,000 total) | ❌ NO |
| Minority class < 100 samples | ❌ NO |
| Need maximum accuracy | ⚠️ Try other methods first |

## Syntax of Sample Method of Pandas


## 🔹 What is `.sample()`?

> Used to **randomly select rows (or columns)** from a DataFrame or Series.

👉 Core use cases:
- Undersampling / Oversampling  
- Shuffling data  
- Creating random subsets  

---

## 🔹 Basic Syntax

```python
df.sample(n=None, frac=None, replace=False, random_state=None, axis=None)
```

---

| Parameter      | Meaning                    | Intuition                                 |
| -------------- | -------------------------- | ----------------------------------------- |
| `n`            | Number of rows to select   | “Give me exactly N random rows”           |
| `frac`         | Fraction of data to select | “Give me % of data”                       |
| `replace`      | Allow duplicates or not    | “Can I pick the same row multiple times?” |
| `random_state` | Fix randomness             | “Make results reproducible”               |
| `axis`         | 0 = rows, 1 = columns      | “Sample rows or columns?”                 |


````python
# fake dataset
data = {
    "feature": np.random.randn(110),
    "target": [0]*100 + [1]*10
}

"""
[0]*100 → 100 zeros
[1]*10 → 10 ones
Combine → [0,0,0,...,1,1]
"""

df = pd.DataFrame(data)
df
````

````python
# NOw Separate Classes

df_majority = df[df["target"] == 0]
df_minority = df[df["target"] == 1]

print( df_majority.head())
print()
print( df_minority.head())
````

````python
# Now Under sample it 

df_majority_sampled = df_majority.sample(n=len(df_minority), random_state=42)

"""
--->> df_majority.sample(...)

        Picks random rows from majority

--->>   n=len(df_minority)   as len(df_minority) == 10

    so pick only 10 rows from majority

---->> random_state=42
        
        Fixes randomness

intuition: “From 100 majority samples, randomly pick 10 → match minority size”
"""

# merge them 
df_balanced = pd.concat([df_majority_sampled, df_minority])

## Shuffle Data of balanced set

df_balanced = df_balanced.sample(frac=1 , random_state=42)

"""
-->> sample(frac=1)

frac=1 → take 100% of data
BUT in random order
    
"""

df_balanced["target"].value_counts()
# BOth zero and one will be 10
````

# 2. Over Sampling

## What is Oversampling?

**Simple definition:** Oversampling means increasing the minority class so both classes become balanced.

- Oversampling means **duplicating or creating new examples** of your minority class (the small group) to match the majority class.

**Analogy:** You have 10 skulls and 100 cookies. You photocopy the 10 skulls 10 times each → now you have 100 skulls and 100 cookies. Balanced!

---

## Types of Oversampling

| Type | How It Works | Best For |
|------|--------------|----------|
| **Random Oversampling** | Randomly duplicates existing minority samples | Quick baseline, simple cases |
| **SMOTE** | Creates synthetic samples between existing minority points | Most common, avoids exact copies |
| **Borderline-SMOTE** | Creates samples only near class boundary | When boundary is unclear |
| **ADASYN** | Creates more samples for harder-to-learn minority points | When minority has varying difficulty |
| **SVMSMOTE** | Uses SVM to find support vectors, creates samples there | Complex boundaries |
| **K-Means SMOTE** | Clusters data, applies SMOTE per cluster | When minority has multiple subgroups |

---

---

## Advantages ✅

| Advantage | Why It Matters |
|-----------|----------------|
| **No information loss** | Keeps ALL original data |
| **Works for small datasets** | Doesn't throw anything away |
| **Improves minority recognition** | Model sees more rare examples |
| **SMOTE creates variety** | Synthetic samples add diversity |
| **Simple to understand** | Just copy or interpolate |
| **Good for text/images** | Can augment creatively |

---

## Disadvantages ❌

| Disadvantage | Why It Hurts |
|--------------|--------------|
| **Overfitting risk** | Model memorizes duplicated samples |
| **Slower training** | More data = longer training |
| **More memory usage** | Larger dataset |
| **Synthetic may be unrealistic** | SMOTE can create impossible examples |
| **No new information** | Just reusing existing data (random oversampling) |
| **Can amplify noise** | Duplicates noisy samples too |

---

## Oversampling vs Undersampling (Quick Comparison)

| Aspect | Oversampling | Undersampling |
|--------|--------------|---------------|
| **What you do** | Add minority samples | Remove majority samples |
| **Dataset size** | Gets LARGER | Gets SMALLER |
| **Training speed** | Slower | Faster |
| **Information loss** | None | Yes (throws data away) |
| **Overfitting risk** | High (duplication) | Low |
| **Best when** | Small dataset | Huge dataset |
| **Memory usage** | Increases | Decreases |

---

## When to Use Oversampling

| Situation | Use Oversampling? | Why? |
|-----------|-------------------|------|
| Small dataset (< 10,000 total) | ✅ YES | Can't afford to lose data |
| Minority class < 500 samples | ✅ YES | Needs more examples |
| You have limited data | ✅ YES | No information to waste |
| Huge dataset (millions) | ⚠️ Maybe | Will be slow |
| Clean, noise-free data | ✅ YES | Won't amplify noise |
| Noisy data | ❌ NO | Will duplicate noise |

---

## SMOTE Explained Simply

**SMOTE** = Synthetic Minority Over-sampling TEchnique

**How it works:**
1. Pick a minority sample (skull A)
2. Find its nearest minority neighbor (skull B)
3. Draw an imaginary line between them
4. Create a NEW skull somewhere on that line
5. Repeat until balanced

**Analogy:** If you have two real friends, SMOTE creates new "friends" who are a mix of both!



````python
np.random.seed(42)

n = 1000

df = pd.DataFrame({
    "monthly_usage": np.random.gamma(2, 20, n),
    "support_calls": np.random.poisson(2, n),
    "session_time": np.random.normal(15, 5, n)
})

# Imbalanced target
df["churn"] = [0]*850 + [1]*150

df = df.sample(frac=1, random_state=42).reset_index(drop=True)
df["churn"].value_counts()
````

````python
#1. Split Classes
df_majority = df[ df["churn"] == 0]
df_minority = df[df["churn"] == 1]

#2 Oversample Minority to migority size

df_minority_oversampled = df_minority.sample(n=len(df_majority), replace=True, random_state=42)
"""
-->> replace = True: “You are allowed to pick the same row again”

Original minority:
A B C

After oversampling:
A B C B A C A A ...
"""

#3. Combine Datasets
df_balanced = pd.concat([df_majority, df_minority_oversampled])

#4. Shuffle DataSet
df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

df_balanced["churn"].value_counts()
````

# 🧠 Random Oversampling & Undersampling — Basic 4 Steps

These are the **core steps used in almost every imbalance handling pipeline**. The idea is simple:  
we either **reduce the majority class (undersampling)** or **increase the minority class (oversampling)**.

---

# ⚖️ 1. Undersampling (Majority Class Reduction)

## 🔹 Step 1: Separate Classes
Split dataset into majority and minority classes.

```python
majority = df[df["target"] == 0]
minority = df[df["target"] == 1]
```

👉 Purpose: We need independent control over each class.

---

## 🔹 Step 2: Reduce Majority Class to Match Minority
Randomly select samples from majority class equal to minority size.
```python
majority_underSampled = majority.sample(n=len(minority), random_state=42)
```


👉 Purpose: Balance the dataset by shrinking the dominant class.

---

## 🔹 Step 3: Combine Data
Merge reduced majority + original minority.
```python
df_balanced = pd.concat([majority_underSampled, minority])
```

👉 Purpose: Reconstruct a balanced dataset.

---

## 🔹 Step 4: Shuffle Data
Randomize order of samples.

```python
majority_underSampled = majority.sample(n=len(minority), random_state=42)
```

👉 Purpose: Prevent ordering bias during training.

---

# 🔁 2. Oversampling (Minority Class Increase)

## 🔹 Step 1: Separate Classes
Split dataset into majority and minority classes.
```python
majority = df[df["target"] == 0]
minority = df[df["target"] == 1]
```


👉 Purpose: Isolate the rare class for augmentation.

---

## 🔹 Step 2: Increase Minority Class to match Majority
Randomly sample minority data **with replacement** until it matches majority size.
```python
minority_overSampled = minority.sample(
    n=len(majority),
    replace=True,
    random_state=42
)
```

👉 Purpose: Duplicate minority patterns to balance learning.

---

## 🔹 Step 3: Combine Data
Merge majority + expanded minority dataset.

```python
df_balanced = pd.concat([majority_OverSampled, majority])
```

👉 Purpose: Build a balanced dataset.

---

## 🔹 Step 4: Shuffle Data
Randomize dataset order.
```python
df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)
```
👉 Purpose: Ensure model does not learn sequence bias.

---

# 🧠 Key Idea Summary

- **Undersampling** → removes majority data  
- **Oversampling** → duplicates minority data  
- Both aim to solve **class imbalance problem**  
- Final goal → balanced learning signal for ML models
