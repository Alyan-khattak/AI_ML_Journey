# Working on anime_messy_dataset.csv
- we will filter it 
- clean it 
- Handle missing values and fill them 
- do Feature Egineering


```python
import pandas as pd
```

```python
df = pd.read_csv("anime_messy_dataset.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
# missing Vales and its percentage
missing_values = df.isnull().sum()
missing_prcnt  = ( missing_values / len(df)) * 100
missing_df = pd.DataFrame( 
    {
        "Missing Values" : missing_values,
        "Missing Percentage": missing_prcnt 
    }
)
print(missing_df.sort_values(by="Missing Percentage", ascending=False))
```

# Stanadardize Col Names

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ","_")
```

```python
df.info()
```

# removing Duplicate Rows

```python
print("Before:", df.shape)

df = df.drop_duplicates()

print("After:", df.shape)
```

# Fix Wrong Data Types of Col
- like age in str , so convert it to numeric

```python
df["rating"] = pd.to_numeric(df["rating"], errors="coerce")
df["episodes"] = pd.to_numeric(df["episodes"], errors="coerce")
```

```python
df.info()
```

#  Checking if a col is takking Unneccessary Space


```python
df["episodes"] = df["episodes"].astype("Int16")
df["rating"] = df["rating"].astype("float16")

catigory_cols = [
    "anime_title",
    "genre"
]

for col in catigory_cols:
    df[col] = df[col].astype("category")
```

```python
df.info()
```

# Handle Missing Values

```python
import numpy as np
df.loc[ df["episodes"] <= 0 , "episodes"] = np.nan
df.loc[ (df["rating"] <= 0  ) & (df["rating"] > 10 ), "rating"] = np.nan
```

# Droping Rows and Col

```python
treshold = len(df) * 0.50 # keep cols that have 85% non mssg values
df.dropna(axis=1, thresh=treshold, inplace=True)
```

```python
df.info()
```

# fill missing Values

```python
df["rating"].fillna(df["rating"].median(), inplace=True)
df["episodes"].fillna(df["episodes"].median(), inplace=True)

df["genre"].fillna(df["genre"].mode()[0], inplace=True)
df["anime_title"].fillna("Unknown", inplace=True)
```

```python
df.isnull().sum()
```

# Now Handling From col_0 to col_44
- they are messiet cols
- they have no clear datatype or as they are mixed type

```python
# Inspect Cols
for col in df.columns:
    if col.startswith("col_"):
        print(col, df[col].dtype, df[col].isnull().sum())
```

# Convert to Numeric Cols 

```python
numeric_cols = [f"col_{i}" for i in range(45)]

for col in numeric_cols:
    # convert to numeric, bad values become NaN
    df[col] = pd.to_numeric(df[col], errors="coerce")
    
    # impossible values: <0 are invalid
    df.loc[df[col] < 0, col] = np.nan
    
    # fill missing with median (robust to outliers)
    df[col].fillna(df[col].median(), inplace=True)
    
    # optimize memory
    df[col] = df[col].astype("float32")
    
```

# Convert Col That Should be catagorical

```python
df.info()
```

```python
for col in numeric_cols:
    if df[col].nunique() < 50:  # heuristic: small number of unique values → categorical
        df[col] = df[col].astype("category")
        print(f"{col} converted to category ")
```

```python
# Remove any remaining duplicates
df = df.drop_duplicates()

# Optional: reset index
df = df.reset_index(drop=True)
```

# Fill Missing Vales

```python
for col in numeric_cols:
    if df[col].dtype.name == "float32":
        df[col].fillna(df[col].median(), inplace=True)
    elif str(df[col].dtype) == "category":
        df[col].fillna(df[col].mode()[0], inplace=True)
```
