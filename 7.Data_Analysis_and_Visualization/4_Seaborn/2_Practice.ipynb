```python
# Sample attendance data
data = pd.DataFrame({
    "Student": ["Alice", "Alice", "Alice", "Bob", "Bob", "Bob", "Charlie", "Charlie", "Charlie"],
    "Subject": ["Math", "Physics", "Chemistry", "Math", "Physics", "Chemistry", "Math", "Physics", "Chemistry"],
    "Attendance": [8, 9, 7, 6, 5, 7, 9, 8, 10]
})


df = data.pivot(index='Student', columns="Subject", values='Attendance')
sns.heatmap(df, annot=True)
```

# Categorical Plots categorica x Numerical

```python
df = pd.DataFrame({
    "Day": ["Mon","Mon","Tue","Tue","Wed","Wed","Thu","Thu","Fri","Fri"],
    "Sales": [100, 120, 90, 95, 130, 125, 80, 85, 110, 115]
})

sns.stripplot(data=df, x="Day", y="Sales")
plt.title("Basic Strip Plot")
plt.show()
```

```python
sns.stripplot(data=tips, x="day", y="total_bill")
```
