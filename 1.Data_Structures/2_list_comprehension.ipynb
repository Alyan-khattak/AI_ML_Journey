```python
### BASIC SYNTAX -->> [expression for item in iterable] = [ i**2 for i in range(10)] 
### expression is the value to be added to list (what operation to be performed on item) 
## item is the variable that takes the value of the item in the iterable on each iteration 
## iterable is a sequence (like list, tuple, string) that we can iterate over 
list = [i**2 for i in range(10)] 
print(list) 




## CONDITION IN LIST COMPREHENSION --->> [expression for item in iterable if condition] -->> [i**2 for i in range(10) if i%2 == 0] 

list2 = [i**2 for i in range(10) if i%2 == 0] 
print(list2) 

even = [i for i in range(20) if i%2 == 0] 
print(even) 

odd = [i for i in range(20) if i%2 != 0] 
print(odd) 

names = ["alyan", "aroosa" , "adil", "asad", "sana"] 
short_names = [name for name in names if len(name) < 5] 
print(short_names) 

long_names = [name for name in names if len(name) >= 5] 
print(long_names) 

uppercase_names = [name.upper() for name in names] 
print(uppercase_names) 

lowercase_names = [name.lower() for name in names] 
print(lowercase_names)
```
