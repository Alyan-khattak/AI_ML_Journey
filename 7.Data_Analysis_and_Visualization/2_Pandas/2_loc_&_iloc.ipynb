# loc and iloc
| Method | How it works              | Think of it as |
| ------ | ------------------------- | -------------- |
| `loc`  | Uses **labels**           | Name lookup    |
| `iloc` | Uses **position numbers** | Index position |


    series.loc[label]

    series.iloc[position]

- in the sq bracket [] we can use fancy indexing and Slicing ALso 

```python
import pandas as pd

power = pd.Series(
    [95, 94, 80, 92],
    index=["Naruto", "Sasuke", "Sakura", "Kakashi"]
)

print(power)
```

```python
# loc[]

print ( power.loc["Naruto"]) 
print ( power.loc["Sakura"] )
print()
print( power.loc[["Naruto" , "Kakashi"]])
```

```python
#iloc
print ( power.iloc[0] ) # althogh the indexes are custom but still we can access them using 0-based via iloc[position]
print( power.iloc[3])
print()
print( power.iloc[[0,1,3]])
print()

# slicing
print( power.iloc[1:3])
print( power.idxmax()) # gives Hghiest Values index
```

# Why loc and iloc
- we can use normal indexing like power[1], power[2] so why use this

- as we have given our custom indexes so using 0-based indexes will give error .. so thats why loc and iloc is used so that we can access it using both position wise and name wise

```python
power[0]
power[2]
```

```python
power.iloc[0]
power.iloc[2]
```
