## 1. WSGI: Web Server GateWay Interface:

> A standard that defines how Python web apps talk to web servers

- Suppose You have a *Web server ( aws )* and A *Web App* (creaed using Python ) that is deployed in the Web Server 

- A request from User comes to Web_server .. the Web server Need to Talk to web-App and redirect the request to the web-app and alo get the reponse back.

> request can be: APIs, /home, some functional 

> IN order to do this communication we use a protocol called WSGI.


# 🧠 WSGI Architecture (Flask)
```python 
User (Browser)
      │
```
