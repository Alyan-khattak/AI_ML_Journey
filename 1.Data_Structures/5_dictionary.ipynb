```python
#accessing values 
name = {"name": "Adil" , "age" : 21}
print(name["name"])
print(name["age"])

#accesing using default get method   -->> .get(key)    ++++    .get(key, msg)
print(name.get("name"))
print(name.get("age"))

print(name.get("dept"))  #this key doesnt exist so will print none

print(name.get("dept" , "Dept not avaible"))  #we can aslo give msg if key doesnt exist -- > .get(key, msg)
print(name.get("roll_no" , "Roll nmbr key doesnt exist"))
```

```python
# Modifying Dictionary elements 
# we can add update and delete elements 

stndt = {"name" : "Alyah" , "age": 21 , "grade": "F"}
print(stndt)

## Adding an element -->>  dict[key] = value 
stndt["Roll_no"] = "009"
stndt["dept"] = "SE"
print(stndt)   # new  key and value will be added 

# Delete key and Value Pair -->>  del dict[key]

del stndt["Roll_no"]
print(stndt)

del stndt["dept"]
print(stndt)

del stndt["grade"]
print(stndt)

del stndt["age"]
print(stndt)

stndt["age"] = 21
print(stndt)

stndt["grade"] = "F"
print(stndt)


```

```python
# Dictionary Methods 
# geting Keys -->>  .key()

dict1 = {"name": "Aroosa" , "age" : 21 , "dept" : "SE"}
print(dict1)

keys = dict1.keys()
print(keys)

#geting values -->> .values()

values = dict1.values()
print(values)

## geting key values pairs  -->> .item()

items = dict1.items()
print(items)



### applying methods
dict1["Roll_no"] = "none"
dict1["Hobbies"] = "Movies"
dict1["uni"] = "COMSATS"

keys= dict1.keys()
print(keys)

items = dict1.items()
print(items)

del dict1["Roll_no"]
del dict1["Hobbies"]
del dict1["uni"]

items = dict1.items()
print(items)





```

```python
#Deep Copy  --->> also inherits the changes done later after copying 
dict2 = {"name": "Aroosa" , "age" : 21 , "dept" : "SE"}

dict2_copy = dict2  # copy dictionary

print(dict2)
print(dict2_copy)

#lets update one key of dict2

dict2["name"] = "Sana"
print("aftr Update dict2" , dict2)
print("after Update copy" , dict2_copy)  #this will also get updated , but we only change dict2 

dict2["age"] = 19
print(dict2)
print(dict2_copy)

dict2["name"] = "a;yan"
print(dict2)
print(dict2_copy)
    
```

```python
# Shallow Copy -- >>> dict_copy =  dict.copy()   ( solves the above problm)
dict3 = {"name": "Aroosa" , "age" : 21 , "dept" : "SE"}
dict3_copy = dict3.copy()

print(dict3)
print(dict3_copy)

dict3["name"] = "sane"   # now only dict3 will change not its copy
print("dict3 after update" , dict3)
print("dict3 copy atfr update ",dict3_copy)


```

```python
#Iteration Over Dictionary --> can iterate over keys , values and items

# iteration over Keys

dict4 = {"name" : "Alyan" , "age" : 93 , "dept" : "SE"}
for keys in dict4.keys():
    print(f"keys : {keys}")

print("\n")
#Over values

for values in dict4.values():
    print(f"values : {values}")

print("\n")

#Over Items

for items in dict4.items():
    print(f"items : {items}")

print("\n")

# Over key, value pair  ( using items as it gives us key,value pair)

for key,value in dict4.items():
    print(f"key: {key} , value : {value}")

print("\n Adding more pairs \n")

# usig methods to modify dict4
dict4["uni"] = "Comsats"
dict4["hobby"] = "charas"
dict4["Aura"] = "-ve"


for key,value in dict4.items():
    print(f"key : {key} , value: {value}")



```

```python
#Nested  Dictionary 
Ninja = {
    "ninja1": {"name": "Kakashi" , "clan": "Hatake" , "jutsu": "chidori"},
    "ninja2": {"name": "Naruto" , "clan": "Uzumaki" , "jutsu": "rasengan"},
    "ninja3": {"name": "Sasuke" , "clan": "Uchiha" , "jutsu": "chidori"}
}

# Accessing its elements 
print(Ninja["ninja1"])
print(Ninja["ninja2"])

print(Ninja["ninja1"]["name"])
print(Ninja["ninja2"]["name"])
print(Ninja["ninja3"]["jutsu"])

#using methods
Ninja["ninja4"] = {"name": "Madara" , "clan" : "Uchiha" , "jutsu" : "Susano"}

print(Ninja["ninja4"])

print("\n")

##------------------------------------------
### Iterating over nested Dictionary
##-----------------------------------------

ninja_items = Ninja.items()
print(f"Ninja Items : {ninja_items}")

print("\n Now Iterations \n")

for ninja_id, ninja_info in Ninja.items():   ## here ninja_id = key , ninja_info = value ( as items gives us key value pairs)
    print(f"Ninja ID : {ninja_id} , Ninja Data : {ninja_info}") 

print("\n Iterations but accessing Nested elements too \n")

for ninja_id,ninja_info in Ninja.items():
    print(f"Ninja Id : {ninja_id} , Ninja Info : {ninja_info}")
    print("\n")
    for key, value in ninja_info.items():   # as Ninnja_info Contains the nested dictionary we can use items on it 
        print(f"key : {key} , value : {value}")
        
```
