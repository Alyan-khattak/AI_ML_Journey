# Filling Missing Values

- Fill Missing Values:

    - it depends on Datatype

            df["age"].fillna(df["age"].mean())  -->> for values use median/mean etc

            df["city"].fillna(df["city"].mode()[0]) -->> for catigorical use mode() ( most common values)


            df["review_text"].fillna("No review")  -->> for text use place holders

```python
# Owner age 
df["owner_age"] = df["owner_age"].fillna(df["owner_age"].median())
```

```python
# Owner income 
df["owner_income"] = df["owner_income"].fillna(df["owner_income"].median())
```

```python
# Daily Waliks per minutes
df["daily_walk_minutes"] = df["daily_walk_minutes"].fillna(
    df["daily_walk_minutes"].median()
)
```

```python
# barks per Day
df["barks_per_day"] = df["barks_per_day"].fillna(df["barks_per_day"].median())
```

```python
df["vet_visits_per_year"] = df["vet_visits_per_year"].fillna(df["vet_visits_per_year"].median())
```

```python
# as differnt breed have different sizes .. global mean would be wrong
# so we group them by breed then in find mean of each group
# Labrador → fill using Labrador median
# Bulldog → fill using Bulldog median

df["weight_kg"] = df.groupby("breed")["weight_kg"].transform( lambda x : x.fillna(x.median()))
```

```python
df["height_cm"] = df.groupby("breed")["height_cm"].transform( lambda x : x.fillna(x.median()))
```

```python
# Categorical Cols are fill with most frequent category

df["diet_type"] = df['diet_type'].fillna(df["diet_type"].mode()[0])
```

```python
df["color"] = df["color"].fillna(df["color"].mode()[0])
```

```python
# there are many sensor col so handle altogather
sensor_col = [col for col in df.columns if "sensor_metric" in col]

for col in sensor_col:
    df[col] = df[col].fillna(df[col].mean())
```

# Feature Engineering

1. Features must add information (not noise)

```python
# Create heavy_dog col
df["is_heavy_dog"] = df["weight_kg"] > 35
# create bmi_like col
df["bmi_like"] = df["weight_kg"] / ( df["height_cm"] / 100 ) **2
# create activity Core col
df["activity_core"] = df["daily_walk_minutes"] * df["age_years"]
# high income owner
df["high_income_owner"] = df["owner_income"] > df["owner_income"].median()
```
