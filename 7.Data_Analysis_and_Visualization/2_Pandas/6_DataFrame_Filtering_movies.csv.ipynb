# filtering Movies.csv

```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("/home/aizen/AI_ML/7.Data_Analysis_and_Visualization/2_Pandas/movies.csv")
movies
```

```python
# Drama Movies
movies[movies["genres"].str.contains("Drama")]
```

```python
# Movies with Runtime Between 100–120 and Gener Action
# as runtime is stored as string so convert to number 1st
movies["runtime"] = pd.to_numeric(movies["runtime"], errors="coerce")
movies[ (movies["genres"].str.contains("Action")) & (movies["runtime"] >= 100) & (movies["runtime"] <= 120) ]
```

```python
# movies with unknown run time
movies[movies["runtime"].notnull()]
```

```python
# top 10 most voted movies
movies.sort_values("imdb_votes", ascending=False).head(10)
```

```python
# Top 10 movies by votes in Comedy genre
movies [ movies['genres'].str.contains("Comedy")].sort_values("imdb_votes", ascending=False).head(10)
```

```python
#Top 10 Action movies after 2010 with rating > 7.5 and runtime < 150


## as runtime is stored as string so convert to number 1st
movies["runtime"] = pd.to_numeric(movies["runtime"], errors="coerce")
movies["release_date"] = pd.to_numeric(movies["runtime"], errors="coerce")
movies[ 
    (movies["genres"].str.contains("Action")) &
    (movies["release_date"] > 2010) &
    (movies["imdb_rating"] > 7.5) &
    (movies["runtime"] < 150)
 
       ].sort_values("imdb_rating").head(10)
```

```python
# Movies with votes above dataset average
movies[movies["imdb_votes"] > movies["imdb_votes"].mean()].head(2)
```

```python
# Movies where title contains "Love"

movies[movies["title_x"].str.contains("Love")].sort_values("title_x").head()
```

```python
# Movies that are NOT Comedy but rating > 8
movies[ ~(movies["genres"].str.contains("Comedy") ) & (movies["imdb_rating"] > 8)].head(2)
```
