### Exception Handling ###

    syntax  -->> try:
                    risky code  ( where exception may arise)
                except:
                    handle error

Events that disrupt normal flow of program.. Occur during runtime.
we need to handle it gracefully and take actions without stopping the execution of program

* if-else handles expected conditions
* try-except handles unexpected runtime errors

### Common Exceptions##

| ----------------- | ------------------------------------------ |
| ValueError        | Wrong type conversion (e.g., `int("abc")`) |
| ZeroDivisionError | Division by zero                           |
| FileNotFoundError | File does not exist                        |
| IndexError        | List index out of range                    |
| KeyError          | Dictionary key missing                     |
| NameError         | Variable not defined                       |
| TypeError         | Wrong data type used in operation          |



    all these are derived of main Class "Exception"
* every exception is derived from this particular class
* when you dont know the name of exception just use above exception class
* always inlude it in normal exceptions at bottom

```python
# Create a function withdraw(balance, amount) that:

def withdraw(balance,amount):
    if amount < 0:
        raise ValueError("Withdrawl amount cant be negative")

    if amount > balance:
        raise ValueError("Insufficeint Funds")

    return balance - amount

balance = 1000

try:
    amount = float(input("enter Withdrawl Amount : "))

    new_balance = withdraw(balance, amount)
except ValueError as ex:
    print("transaction Error : ",ex)

else: # Runs ONLY if no exception occurred
    print("Withdrawl Successful")
    print("Remaing Balance : ",new_balance)

finally:
    print("transaction Completed")
```

```python
# students score processor

students = {
    "Alyan": [80, 90, 70],
    "Aroosa": [100, 20, 90],
    "Shayan": [50, 60, 100]
}


def validate_scores(scores):
    # Check each score
    for score in scores:
        if score < 0 or score > 100:
            raise ValueError(f"Invalid score detected: {score}")


try:
    # Step 1: Validate all student scores
    for name, scores in students.items():
        validate_scores(scores)

    # Step 2: Calculate averages using map
    averages = dict(
        map(
            lambda item: (item[0], sum(item[1]) / len(item[1])),
            students.items() # items return key,value pair in form of tple ("Alyan", [80, 90, 70]) 
                             # item[0] is key , item[1] is whole list of values and len(item[1]) each value in item[1]
        )
    )

except ValueError as e:
    print("Validation Error:", e)

else:
    # Step 3: Filter students with average >= 60
    passed = dict(
        filter(lambda item: item[1] >= 60, averages.items())
    )

    print("Passed Students:", passed)

finally:
    print("Processing completed.")
```

```python
inventory = {
    "laptop": (5, 1000),
    "phone": (0, 500),
    "tablet": (3, 300)
}


def validate_inventory(item, data):
    quantity, price = data  # as data is tple (0,100) -->> unpacking a tpl

    if quantity < 0:
        raise ValueError(f"Negative quantity for {item}")

    return (item, quantity, price)


try:
    # Step 1: Validate using map
    validated = list(
        map(lambda item: validate_inventory(item[0], item[1]), inventory.items())
    )  # Each element is a tuple (key, value) where value is another tuple (quantity, price).

except ValueError as e:
    print("Inventory Error:", e)

else:
    # Step 2: Remove zero quantity using filter
    available = list(
        filter(lambda item: item[1] > 0, validated)
    )

    # Step 3: Calculate total value
    total_value = sum(
        map(lambda item: item[1] * item[2], available)
    )

    print("Available Inventory:", available)
    print("Total Inventory Value:", total_value)

finally:
    print("Inventory check complete.")
```
