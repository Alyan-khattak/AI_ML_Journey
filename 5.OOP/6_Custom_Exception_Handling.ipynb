## Raise ##
    “STOP the program right here and throw an error.”

* Creates an exception object

* Stops current execution

* Looks for a matching except block

* If found → jumps there

* If not found → crashes program

* After raise, NOTHING below it runs.

```python
# Custom Exception

class AgeTooSmallError(Exception):
    pass


age = 10

if age < 18:
    raise AgeTooSmallError("You Must be 18+")
```

```python
## Login System
## Custom Exceptions
class InvalidPinError(Exception):
    pass

class AccountBlockedError(Exception):
    pass


class ATM:
    def __init__(self, pin):
        self.pin = pin
        self.attemps = 3

    
    def login(self, entered_pin):
        if self.attemps == 0:
            raise AccountBlockedError("Account is Blocked")

        if entered_pin != self.pin:
            self.attemps -= 1
            raise InvalidPinError(f"Wrong PIN. Attemps Left: {self.attemps}")
        

        print("Login Sccuessfully!")



atm = ATM("1234")

try:
    atm.login("9792")
except InvalidPinError as ex:
    print(ex)
except AccountBlockedError as ex:
    print(ex)
except Exception as ex:
    print(ex)
```

```python
## Hospital Sytem 
## custom exception

class AgeRestrictionError(Exception):
    pass

class InvalidPatientIDError(Exception):
    pass

class Hospital:

    def admit(self, patient_id, age):
        if not isinstance(patient_id, int):
            raise InvalidPatientIDError("Patient ID must be intiger")
        
        if age < 0:
            raise AgeRestrictionError("Age Can not be negative")
        
        print("Patient Admitted")

h1 = Hospital()
h1.admit("32",90)
```

```python
class GameError(Exception):
    pass

class PlayerDeadError(GameError):
    pass

class LevelLockedError(GameError):
    pass

class InsufficientEnergyError(GameError):
    pass


## Game class
class Game:
    def __init__(self, alive = True, energy = 100, unlocked_lvls = None):
        self.alive = alive
        self.energy = energy
        self.unlocked_lvls = unlocked_lvls

    def play_lvl(self, lvl):
    
        if not self.alive:
            raise PlayerDeadError("player is Dead")
    
        if lvl not in self.unlocked_lvls:
            raise LevelLockedError("Level is Blocked")
    
        if self.energy < 20:
            raise InsufficientEnergyError("Not enough engery ")
    
        self.energy -= 20
        return f"Playing level {lvl}. Engergy left {self.energy}"


##
game = Game(alive = True, energy = 15, unlocked_lvls = [1,2])

try: 
    rlst = game.play_lvl(2)
except PlayerDeadError as ex:
    print(ex)
except LevelLockedError as ex:
    print(ex)
except GameError as ex:
    print(ex)
else:
    print(rlst)
finally:
    print("Game Action Completed")

```

```python
## Custome exceptions
class PlatformError(Exception):
    pass

class AuthenticationError(PlatformError):
    pass

class EnrollmentError(PlatformError):
    pass

class SubmissionError(PlatformError):
    pass

class PermissionError(PlatformError):
    pass


### Business Classes

class User:

    def __init__(self, name, logged_in = False):
        self.name = name
        self.__logged_in = logged_in #private

    def login(self):
        self.__logged_in = True

    def logout(self):
        self.__logged_in = False

    def Check_Authentication(self):
        if not self.__logged_in:
            raise AuthenticationError("User not logged in")
        

## class student ( single inheritance)

class Student(User):

    def __init__(self, name, logged_in = False):
        super().__init__(name, logged_in) # constructor Overriding
        self.course = []

    def enroll(self, course):

        self.Check_Authentication()

        if course in self.course:
            raise EnrollmentError("Already Enrolled")
        
        self.course.append(course)
        return f"{self.name} enrolled in {course}"
    

# Multilvl inheritance
class PremiumStudent(Student):

    def __init__(self, name, logged_in=False):
        super().__init__(name, logged_in)

        self.premium_access = True

    #Method Overiding
    def access_preium_content(self):

        self.Check_Authentication()

        if not self.premium_access:
            raise PermissionError("Premium Access Required")
        
        return "Acessing Premium Content"

  # Single INheritance  
class Instructor(User):  # here student is objct of Student class so Aggregation
    def grade_asnmnt(self, student, score):

        self.Check_Authentication()

        if not isinstance(student, Student):
            raise SubmissionError("Invalid Student Object")
        
        if score < 0 or score > 100:
            raise SubmissionError("Score must be between 0 and 100")
        
        return f"{student.name} graded with score {score}"


#Multiple Inheritance
class AdminInstructor(Instructor, Student):

    def __init__(self, name, logged_in = False):
        super().__init__(name, logged_in)

    def delete_course(self, course):
        self.Check_Authentication()
        return f"Course {course} deleted"
    
## Execution

premium = PremiumStudent("Alyan")
tchr = Instructor("Riaz Ahmad")

try:
    # accessing without login
    print(premium.access_preium_content())

except AuthenticationError as ex:
    print("Platform Error : ",ex)

    # fixed eror and retried ( nested try-excep)
    premium.login()

    try:
        print(premium.access_preium_content())
    except PlatformError as ex:
        print("Platform Error : ",ex)

except PlatformError as ex:
    print(ex)
else:
    print("no Errors")
finally:
    print("operation Finished")


try:
    premium.enroll("Python Mastery")
    premium.enroll("Python Mastery")  # Duplicate → error

except EnrollmentError as e:
    print("Enrollment Problem:", e)

    # Re-raise after logging
    raise

except PlatformError as e:
    print("Other Platform Issue:", e)

finally:
    print("Enrollment attempt completed.")
```
