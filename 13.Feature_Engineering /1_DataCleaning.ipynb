# Data Cleaning # 

> There are Three Types 

## MCAR: Missing Completely at Random

- Missing values have no pattern at all
- The missing value is random
- Not related to any other columns

> safe to 

    drop rows
    fill with mean/median


## MAR: Missing At Random

- Missing depends on another Column

> Age missing Depends on Gender: Females usually dont fill age 

> Do: 
- group-based filling

## MNAR: Missing Not at Random

- Missing depends on itself (hidden reason)

    High-income people don’t reveal salary
    So high salaries = missing

Dangerous:

    Mean/median filling → wrong

✔ Better:
    
    create new feature:
