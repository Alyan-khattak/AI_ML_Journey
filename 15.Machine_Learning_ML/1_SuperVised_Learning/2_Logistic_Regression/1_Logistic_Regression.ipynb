````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
````

````python
# we will create a Data using sklearn 
# Create Data Set for  Classification problem

from sklearn.datasets import make_classification
````

````python
# Create the Dataset using Make_classification 
X, y = make_classification(n_samples=10000, n_features=10, n_classes=2, random_state= 42)

# n_samples : nmbr of datapoits
# n_features : nmbr of independednt features 
# n_classes : how many classes of output feature ( we select 2 so binary )

# this will return data into section independet X and  Y dependent 
````

````python
X
# see the features are already standardized ( we dont need to implemet scaler on it )
````

````python
pd.DataFrame(X)
````

___
# Train Test Split
___


````python
from sklearn.model_selection import train_test_split
````

````python
X_train, X_test, y_train , y_test = train_test_split(X, y, test_size=0.30 , random_state=42)
````

___
# Model Training 
___

````python
from sklearn.linear_model import LogisticRegression
````

````python
model = LogisticRegression()
model.fit(X_train, y_train)
````

````python
y_pred = model.predict(X_test)
````

````python
print("actual Values: ", y_test)
print("predicted Data : ", y_pred)
````

````python
# this gives us probabilty 
model.predict_proba(X_test)
````

___
# Performance Matrix
___

````python
from sklearn.metrics import accuracy_score , confusion_matrix, classification_report
````

````python
score = accuracy_score(y_test, y_pred)
print(score)
````

````python
# Confusion matrix .. since this is binar classification we'll get  2x2 matrix
confusion_mtrx = confusion_matrix(y_test, y_pred)
print(confusion_mtrx)
````

````python
print(classification_report(y_test, y_pred))
````

# Hyperparameters of Logistic Regression

Hyperparameters are settings that you choose **before training** the Logistic Regression model.

They control how the model learns.

---

# Scikit-Learn Syntax

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(
    penalty="l2",
    C=1.0,
    solver="lbfgs",
    max_iter=100,
    random_state=42
)

model.fit(X_train, y_train)
```

---

# 1. `penalty`

Controls the type of regularization.

```python
penalty="l2"
```

Possible values:

| Value          | Meaning                        |
| -------------- | ------------------------------ |
| `"l2"`         | Ridge Regularization (Default) |
| `"l1"`         | Lasso Regularization           |
| `"elasticnet"` | Elastic Net                    |
| `None`         | No regularization              |

Example:

```python
LogisticRegression(penalty="l1")
```

---

# 2. `C`

The **inverse of regularization strength**.

```python
C=1.0
```

Think of it like this:

```text
Small C
    ↓
More Regularization

Large C
    ↓
Less Regularization
```

Examples:

```python
C=0.01
```

Very strong regularization.

```python
C=1
```

Default.

```python
C=100
```

Almost no regularization.

---

# 3. `solver`

Algorithm used to optimize (train) the model.

```python
solver="lbfgs"
```

Common solvers:

| Solver        | Supports                            |
| ------------- | ----------------------------------- |
| `"lbfgs"`     | L2, Multiclass (Default)            |
| `"liblinear"` | Small datasets, L1 & L2             |
| `"saga"`      | L1, L2, Elastic Net, Large datasets |
| `"newton-cg"` | L2                                  |
| `"sag"`       | Large datasets                      |

Example:

```python
solver="liblinear"
```

---

# 4. `max_iter`

Maximum number of training iterations.

```python
max_iter=100
```

If you see:

```text
ConvergenceWarning:
```

Increase it.

Example:

```python
max_iter=1000
```

---

# 5. `random_state`

Controls randomness.

```python
random_state=42
```

Used for reproducible results.

---

# 6. `class_weight`

Useful for imbalanced datasets.

```python
class_weight="balanced"
```

Example:

```text
Normal Emails : 950

Spam Emails : 50
```

Without balancing:

The model may predict almost everything as **Normal**.

With:

```python
class_weight="balanced"
```

Both classes receive appropriate importance.

---

# 7. `multi_class`

Controls multiclass classification strategy.

Example:

```python
multi_class="ovr"
```

Possible values:

| Value           | Meaning                |
| --------------- | ---------------------- |
| `"ovr"`         | One-vs-Rest            |
| `"multinomial"` | Softmax Classification |

> **Note:** In newer versions of Scikit-Learn, `multi_class` is often selected automatically based on the solver.

---

# Complete Example

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(
    penalty="l2",
    C=1.0,
    solver="lbfgs",
    max_iter=500,
    random_state=42
)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

---

# Most Important Hyperparameters

| Hyperparameter | Purpose                     |
| -------------- | --------------------------- |
| `penalty`      | Type of regularization      |
| `C`            | Strength of regularization  |
| `solver`       | Optimization algorithm      |
| `max_iter`     | Maximum training iterations |
| `class_weight` | Handle imbalanced classes   |
| `random_state` | Reproducible results        |

---

# Rule of Thumb

* **`penalty`** → Choose the regularization method.
* **`C`** → Controls how strong the regularization is.
* **`solver`** → Choose the optimization algorithm.
* **`max_iter`** → Increase if the model doesn't converge.
* **`class_weight="balanced"`** → Use for imbalanced datasets.

---

## One-Line Definition

> Logistic Regression hyperparameters control **how the model learns**, **how much regularization is applied**, and **how the optimization process is performed** before training begins.

````python
from sklearn.linear_model import LogisticRegression
````

````python
model = LogisticRegression()

penalty = ["l1", "l2", "elasticnet"]
c_values = [100,10,1.0,0.1,0.01]
solver = ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']

# above are all the paramters: we will check which one is best for our dataset
````

````python
params = dict( penalty = penalty, C = c_values, solver = solver )
params
````

````python
# Grid Search Cv
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
````

````python
cv = StratifiedKFold()
````

````python
grid = GridSearchCV(estimator=model, param_grid=params, scoring='accuracy', cv = cv, n_jobs= -1)

# grid find the best combination of parameter
````

````python
grid.fit(X_train, y_train)
````

````python
# Now see which paramters are the best 
grid.best_params_

# so output are best parameter
````

````python
grid.best_score_
````

````python
grid.predict(X_test)
````

````python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
````

````python
score = accuracy_score(y_test, grid.predict(X_test))
print(score)
````

````python
confusion_mtrx = confusion_matrix(y_test, grid.predict(X_test))
print(confusion_mtrx)
````
