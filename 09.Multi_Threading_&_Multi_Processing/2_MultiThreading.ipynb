# Multi Threading 
- When to Use 

> I/O Bound Tasks: When ur program spend more time waitin for I/O.

> Cuncurrent Execution: When you want to improve throughput of an Application by performing Multiple Operations 

## Syntax
```python
# create thread
t = threading.Thread(target = function_name)

t.start() # start thread

t.join() # Wait for thread to Finish and join main thread


## if a functon has taker Arguments 

add(a,b)

t = threading.Thread(target = add, args = (5,3) )  # args is Value PAssed into that function
                                                   # add(5,3)
```

---

| Part       | Meaning             |
| ---------- | ------------------- |
| `Thread()` | create thread       |
| `target`   | function to run     |
| `start()`  | starts thread       |
| `join()`   | wait for completion |


````python
import threading
import time
````

````python

# simulate wait for I/O


def print_nums():

    for i in range(5):
        time.sleep(2) # it sleeps for 2 secs before printing next num ( I/O wait simulate )
        print(f"Number: {i}")


def print_letters():

    for i in "ABCDE":
        time.sleep(2)
        print(f"Letter: {i}")


## recording start  and end time
t1 = time.time() # get current time
print_nums()
print_letters()

total_time = time.time() - t1 # gets total_execution time
print(total_time)
````

- Now I want when One function is sleeping I want the Othe Function to COntinue once the sleep ends it comes back to it 

- as of now whole code runs in One Tread 
- create two threads , in each thread run one function
- so now One thread will be responsible for running one function and when wait comes in one thread we switch to another thread

````python
# The Above COde using Threading 

def print_nums():

    for i in range(5):
        time.sleep(2) # it sleeps for 2 secs before printing next num ( I/O wait simulate )
        print(f"Number: {i}")


def print_letters():

    for i in "ABCDE":
        time.sleep(2)
        print(f"Letter: {i}")

# Creatng Threads

t1 = threading.Thread(target=print_nums)
t2 = threading.Thread(target=print_letters)

t = time.time() # note time
# start the threads
t1.start()
t2.start()

# wait for thread to complete 
t1.join()
t2.join()

ttl_time = time.time() - t
print(ttl_time)

# execution tyme is halved
````

````python

def add(a, b):
    print(a + b)

t1 = threading.Thread(target=add, args=(5, 3)) # args is value Passed into that Function if it takes Value 
t1.start()
t1.join()
````

````python
# Fast vs Slow Tasks
def fast_task():
    for i in range(5):
        time.sleep(1)
        print(f"Fast → {i}")

def slow_task():
    for i in range(3):
        time.sleep(2)
        print(f"Slow → {i}")

start = time.time()
t1 = threading.Thread(target=fast_task)
t2 = threading.Thread(target=slow_task)


t1.start()
t2.start()

t1.join()
t2.join()

print("Time:", time.time() - start)
````

````python
# Creating Threads dynamicaly 
# in real work flow we dont create fixed threads

def worker(n):
    print(f"Worker {n} started")
    time.sleep(2)
    print(f"Worker {n} finished")

threads = [] # empty list: it will recive Threads nd become list of theads -> [t,t,t] each t is separate thread

start = time.time()

for i in range(5):
    t = threading.Thread(target=worker, args=(i,))
    threads.append(t)  # append the thread t in list named threads i.e threads = [t,t,t,t,t]
    t.start() # start thread

for t in threads:
    t.join()

print("Time:", time.time() - start)
````

````python
# Simulating Download Manager
import random

def download(file):
    print(f"Downloading {file}...")
    time.sleep(random.randint(1, 4))
    print(f"{file} downloaded")

files = ["file1", "file2", "file3", "file4"]

threads = []

start = time.time()

for f in files:
    t = threading.Thread(target=download, args=(f,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()

print("Total Time:", time.time() - start)
````
