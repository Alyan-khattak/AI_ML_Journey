## Importing Modules ###
It makes code easy and reusable

```python
# importing MOdule
import math
math.sqrt(16)  # from math we are usng method sqrt
```

```python
# Importing Function from a pkg
from math import sqrt,pi
print(sqrt(16))
print(pi)
```

```python
# Importing nmpy and giving it a name
import numpy as np   # np is alias for numpy
np.array([1,2,3,4])
```

```python
# to import all the Methods from a Module use " * "
from math import *
sqrt(16)   # now we can use the methods directly
```
