### Inheritance ###
When One class( child/derived) derives or inherits the properties and methods of another class (parent/base)

    Syntax:     class Parent:  -->> parent class
                    pass

                class Child(Parent):  -->> child class
                    pass

## Types of Inheritance ##

### Single Lvl Inheritance ##
    One parent → One child 

    class A:
        pass

    class B(A):
        pass

### Multi LvL Inheritance ###
    Grandparent → Parent → Child

    class A:
        pass

    class B(A):    -> inherits A
        pass

    class C(B):    -> INherits B
        pass

### Multiple Inheritance ###
    Child inherits from multiple parents.

    class Father:
        def skill1(self):
            print("Business")

    class Mother:
        def skill2(self):
            print("Cooking")

    class Child(Father, Mother):
        pass


### Hierarchical ###
    One parent → multiple children

    class Animal:
        pass

    class Dog(Animal):   --> inherits animal
        pass

    class Cat(Animal):   -->> also inherits animal
        pass



### Super KeyWord ###
super() is used to call methods from the parent class inside a child class.

use to resuse code from parent

* its use is just like we did in java with constructors and methods

    Syntax: ##Mathods##

       -> super().method_name()

        ##Constructor##

      -->  super().__init__(arguments)


### Method Overiding ###
A child class provides its own version of a method that already exists in the parent class.
same name, paramters but different implementation

    class Animal:
        def speak(self):
            print("Animal makes a sound")

    class Dog(Animal):
        def speak(self):   # overriding
            print("Dog barks")

* overiding but still want to use parent method use super()


```python
### Single LvL Inheritance : One Parent → One Child

class Ninja:
    def use_chakra(self):
        print("Ninja uses Chakra")

class Naruto(Ninja):
    def rasengan(self):
        print("Rasengan")


n = Naruto()
n.rasengan()
n.use_chakra()  # i can now use this Too

print()
print()
print("using Super")
## USing Super

class Character:
    def __init__(self, name, health):
        self.name = name
        self.health = health

class Warrior(Character):
    def __init__(self, name, health, weapon): # see we aldy made constructor for name, health above so use that
        super().__init__(name,health) # call parent constructor
        self.weapon = weapon
        print("Warrior Ready")

w = Warrior("Alyan", 78, "Sword")

```

```python
class Shinobi:
    def __init__(self, name):
        self.name = name

    def introduce(self):
        print(f"I am {self.name}")

class Sasuke(Shinobi):
    def __init__(self, name, clan):
        super().__init__(name)
        self.clan = clan

    def show_clan(self):
        print(f"I belong to {self.clan}")

s1 = Sasuke("Sasuke", "Uchiha")
s1.show_clan()
```

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary


    def show_details(self):
        print(f"{self.name} earns {self.salary}")


class Manager(Employee):
    def __init__(self, name, salary, dept):
        super().__init__(name,salary)
        self.dept = dept

    #method Overiding
    def show_details(self):
        super().show_details() # using super to call parent method
        print(f"department : {self.dept}")


m1 = Manager("Aroosa",999999 , "Laziness")
m1.show_details()
        
```

### Muli Level Inheritance ###


```python
class Animal:
    def eat(self):
        print("animal eatinng")

class Mammal(Animal):
    def walk(self):
        print("Mammal Walking")

class Dog(Mammal): # now this class can use methods of parnt and grandparent
    def bark(sef):
        print("dog barking")

d = Dog()
d.bark()
d.eat()
d.walk()
```

```python
class Entity:
    def exist(self):
        print("Entity exists")

class Person(Entity):
    def live(self):
        print("Person lives")

class Employee(Person):
    def work(self):
        print("Employee works")

class Manager(Employee):
    def manage(self):
        print("Manager manages")

m = Manager()
m.exist()
m.live()
m.work()
m.manage()
```

### Multiple Inheritane ###
one child → multiple parents

```python
class Father:
    def skills1(self):
        print("Father : Bussiness Skills")

class Mother:
    def skills2(self):
        print("Mother: Coding Skills")

class Child(Father,Mother): #inherits both father and mother
    pass

c = Child()
c.skills1()
c.skills2()
```

```python
class Device:
    def __init__(self, brand):
        self.brand = brand

    def power_on(self):
        print(f"{self.brand} device powering on...")


class CameraDevice(Device):
    def take_pic(self):
        print(f"{self.brand} takes photo ")


class MusicDevice(Device):
    def play_music(self):
        print(f"{self.brand} plays music")


class SmartPhone(CameraDevice,MusicDevice):
    def __init__(self, brand, os):
        super().__init__(brand)
        self.os = os

    def power_on(self):
        super().power_on()
        print(f"{self.brand} smartphone with {self.os} OS booting")

# Testing
phone = SmartPhone("Samsung", "Android")
phone.power_on()    # overridden + super
phone.take_pic()  # from CameraDevice
phone.play_music()  # from MusicDevice
```
