### File Handling ###
    Open the file
    Do something (read / write / append)
    Close the file

* ->> Text File == .txt ,.doc , .log
* -->> Binary File == .mp4, .img ,.jpg etc
### Modes ###
| Mode   | Meaning           |
| ------ | ----------------- |
| `"r"`  | read              |
| `"w"`  | write (overwrite) |    
| `"a"`  | append            |
| `"x"`  | create new file   |
| `"rb"` | read binary       |
| `"wb"` | write binary      |

## Methods #

* open("file_name", "mode")
* read()         -->> return file content in form of string
* close()  -->> close opened file
* readline() -->> reads one line at a time

* write("new data") -->> writes in file when opened in write mode





```python
f = open("demo.txt", "r")  #we oppen file in f 
data = f.read()    # f.read() will return file content in form of string
print(data)
f.close()
```

```python
f = open("demo.txt" , "r")
line1 = f.readline() # this read one lines at a time
line2 = f.readline()
print(line1)
print(line2)
f.close()
```

```python
# writing a file 
f = open("demo.txt" , "w")
f.write("I am writing this using 'w' mode")
f.close()  # the above will now override in our file

f = open("demo.txt" ,"r")
data1 = f.read()  # the new thing will print
print(data1)  

f = open("demo.txt", "a")
f.write("now I am appending in the file")
f.close() # this line will append

f = open("demo.txt" , "r")
data2 = f.read()
print(data2)
f.close()
```
