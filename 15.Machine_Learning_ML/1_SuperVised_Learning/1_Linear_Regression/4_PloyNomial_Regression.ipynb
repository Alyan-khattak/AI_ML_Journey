# Polynomial Regression

Polynomial Regression is a regression technique used when the relationship between the input features and the target variable is **non-linear**.

Although it produces a curved line, it is still considered a form of **Linear Regression** because the model remains linear in its parameters (coefficients).

---

# Why Does Simple Linear Regression Fail?

Simple Linear Regression assumes that the relationship between the feature and target is a straight line:

$$
\hat{y} = \theta_0 + \theta_1x
$$

Consider the following data:

| X | y  |
| - | -- |
| 1 | 1  |
| 2 | 4  |
| 3 | 9  |
| 4 | 16 |
| 5 | 25 |

The actual relationship is:

$$
y=x^2
$$

---

## Best Fit Line

Simple Linear Regression tries to fit:

```text id="71x6zr"
        *
      *
    *
  *
*
-------------------->
```

A straight line cannot capture the curve properly.

This results in:

* High bias
* Underfitting
* Poor predictions

---

## Actual Relationship

```text id="20g3zl"
            *
         *
      *
   *
*
-------------------->
```

The data follows a curve rather than a straight line.

---

# Intuition

Instead of fitting:

$$
y = \theta_0 + \theta_1x
$$

we create additional features:

$$
x^2,;x^3,;x^4,\dots
$$

and fit:

$$
\hat{y}
=======

\theta_0
+
\theta_1x
+
\theta_2x^2
+
\theta_3x^3
+\dots
$$

This allows the model to learn curved relationships.

---

# Simple Polynomial Regression

Simple Polynomial Regression uses **one feature**.

## Formula

### Degree 2

$$
\hat{y}
=======

\theta_0
+
\theta_1x
+
\theta_2x^2
$$

---

### Degree 3

$$
\hat{y}
=======

\theta_0
+
\theta_1x
+
\theta_2x^2
+
\theta_3x^3
$$

---

| Study Hours | Marks |

| 5           | 130   |

The relationship is curved:

```text id="9qim5v"
               *
            *
         *
      *
   *
-------------------->
```

A polynomial curve fits the data much better than a straight line.

---

# Multiple Polynomial Regression

Multiple Polynomial Regression uses multiple features and their polynomial terms.

Suppose we have:

* Area
* Bedrooms

The model becomes:

$$
\hat{y}
=======

\theta_0
+
\theta_1x_1
+
\theta_2x_2
+
\theta_3x_1^2
+
\theta_4x_2^2
+
\theta_5x_1x_2
$$

Where:

| Symbol   | Meaning          |
| -------- | ---------------- |
| $x_1$    | First feature    |
| $x_2$    | Second feature   |
| $x_1x_2$ | Interaction term |

---

# Degree of Polynomial

| Degree | Shape               |
| ------ | ------------------- |
| 1      | Straight line       |
| 2      | Parabola            |
| 3      | Cubic curve         |
| 4+     | More complex curves |

Examples:

$$
Degree\ 1:
\quad
y=a+bx
$$

$$
Degree\ 2:
\quad
y=a+bx+cx^2
$$

$$
Degree\ 3:
\quad
y=a+bx+cx^2+dx^3
$$

---

# Advantages

* Captures non-linear relationships.
* More flexible than simple linear regression.
* Easy to implement.

---

# Disadvantages

* Can easily overfit.
* Higher degree increases complexity.
* Harder to interpret.

---

# When to Use Polynomial Regression

Use polynomial regression when:

* Scatter plots show a curved relationship.
* Linear regression underfits.
* The relationship between variables is non-linear.

---

# One-Line Definition

> Polynomial Regression extends Linear Regression by adding powers and combinations of features so that curved relationships can be modeled.

````python
X = 6 * np.random.rand(100,1) - 3
Y = 0.5 * X**2 + 1.5*X + 2 + np.random.rand(100, 1)
# Quadratic Equation used: y = 0.5 X^2 + 1.5X + 2 +outliers
plt.scatter(X,Y, color = "Green")
plt.xlabel("X")
plt.ylabel("y")

# see the data has non  linear relation
````

# Polynomial Regession 
___

````python
from sklearn.preprocessing import PolynomialFeatures
````

````python
poly  = PolynomialFeatures(degree=2, include_bias=True)

# degree = 2 means -->> h0X = Bo + b1X1 + B2X2^2
# include Bias means -->> multiple Bo with with 1 
````

````python
X_Train_poly = poly.fit_transform(X_train)  # degree 2 will be created for X-train
X_test_poly = poly.transform(X_test)
````

````python
X_Train_poly
````

___

# Now Use Regression

___

````python
from sklearn.linear_model import LinearRegression
````

````python
model = LinearRegression()
model.fit(X_Train_poly, y_train)
````

````python
y_pred = model.predict(X_test_poly)
````

````python
print("Predicted  :", y_pred[:5].round(2))
print("Actual     :", y_test[:5])
````

````python
from sklearn.metrics import r2_score
````

````python
r2_score  = r2_score(y_test, y_pred)
````

````python
print("score : ", r2_score)
# see our Accuracy reached 98 % 
````

````python
model.coef_
# there are tell valus as B0, B1X1 and B2X2
````

````python
plt.scatter(X_train, model.predict(X_Train_poly)) # best fit line 
plt.scatter(X_train, y_train)
````
