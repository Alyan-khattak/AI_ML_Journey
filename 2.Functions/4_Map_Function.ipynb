### Map FuncTion ### 
Applies a Given Function (add,even_odd etc) to all the items in an iterable (list, tuple etc)
# Take a function and apply it to every item in an iterable #

### Syntax --->>> map(function, iterable) ##
    function -->> our predefined function ( what to do with each item)
    iterable -->> tple , list, set etc

# map() returns a map object , so usually we wrap it in a list ##



```python
# we define a function here 
def square(x):
    return x*x

square(10)
```

```python
# I have a list (iterable) and want to apply the above function on each ele 
numbers = [1,2,3,4,5,6,78,79]
map(square, numbers) #it takes numbers and call sqaure function for each value 

# wrap it in a list 
list(map(square, numbers))
```

```python
# convert string into intiger 
nums = ["1", "2" , "3"]
result = map(int, nums)  #int -->> inbuilt type cast function
print(list(result))
```

```python
# convert String to uppercase
def to_upper(s):
    return s.upper()

names = ["alyan", "adil", "aroosa" , "asad" , "shayan"]

upper = map(to_upper, names)
print(list(upper))
```

```python
# get length 
def get_length(s):
    return len(s)

names = ["alyan", "adil", "aroosa" , "asad" , "shayan"]

list(map(get_length, names))
```

```python
# strip extra space from words  
def clean_string(s):
    return s.strip()

data = ["  alyan ", " khan  ", " khattak "]

print(list(map(clean_string, data)))
```

## lamda with Map ##
As lambda is also a function so we can use it in map  
map(fucntion , iterable) -->> in function we can write lambda -->> map(lamda x:x+2 , iterable)


```python
#lambda with Map  -->> square of each item in a list
nm = [1,2,34,5,6,7,8,9,7]

sqs = map(lambda x: x**2, nm )
print(list(sqs))
```

```python
# Convert strings into intigers
numss = ["1", "2" , "3" , "4"]
list(map(lambda x: int(x) , numss))
```

```python
# Add 5 to every nmbr 
num1 = [1,3,4,5,93,6,8,9]
list(map(lambda x: x+5 , num1))
```

```python
# Add Two lits 
list1 = [1,4,6,7]
list2 = [4,5,98,4]

added_lists = map(lambda a,b: a+b , list1, list2)
print(list(added_lists))
```

```python
# Extract Second Element from a Tupple 
pairs = [(1, 10), (2, 20), (3, 30)]
seconds = map(lambda tpl: tpl[1] , pairs)

print(list(seconds))
```
