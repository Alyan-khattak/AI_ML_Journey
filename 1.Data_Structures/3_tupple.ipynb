```python
# Tuple are ordered and immutable (cannot be changed after creation)
my_tuple = (1, 2, 3, 4, 5)
print(my_tuple)
# Accessing elements in a tuple
print(my_tuple[0])  # Output: 1
print(my_tuple[2])  # Output: 3
# Tuples can contain mixed data types
mixed_tuple = (1, "Hello", 3.14 )
print(mixed_tuple)
```

```python
# Tuple Opertaions

# Concatenation 
tp1 = (1,2,3)
tp2 = (4,5,6)

concatinated_tuple = tp1 + tp2 
print(concatinated_tuple)

# Repetition --> use  * operator to repeat the tuple
repeated_tuple = tp1 * 3 # this will repeat the tuple 3 times
print(repeated_tuple)

repeated_tuple2 = tp2 * 2 # this will repeat the tuple 2 times
print(repeated_tuple2)
```

```python
#packing and unpacking of tuples

# Packing a tuple
                                    
packed_tuple = 1, "alyan" , 2, 3.14  # by default python will pack values into a tuple when we assign multiple values to a single variable
# when many values are assigned to a single variable, it is called packing of tuple
print(packed_tuple)

# Unpacking a tuple
a , b , c , d = packed_tuple # we can unpack the tuple into individual variables
print(a) # 1
print(b) # "alyan"
print(c) # 2
print(d) # 3.14
```

```python
nums = (1,2,3,4,5)
# Unpacking with * operator
first , *middle , last = nums # " * " is used to unpack the remaining elements of the tuple into a list
print(first) # 1
print(middle) # [2,3,4]
print(last) # 5


a , b = nums # this will raise error because there are more values in the tuple than variables to unpack into
print(a) # 1
print(b) # 2
```

```python

# we need to ignore the remaining values in the tuple
# to ignore we use " _ " as a placeholder for the values we want to ignore
# and we can use " * " to ignore the remaining values and unpack only the first value

a , *_ , last = nums # this will unpack the first value into a and the last value into last and ignore the remaining values
print(a) # 1
print(last) # 5

pair = ((1,2), (3,4), (5,6))

for a,b in pair: 
    print(a,b)

# enumerate is a func that ruturns index nd value of the iterable object
items = ( "apple", "banana", "cherry" ) 
for index, item in enumerate(items): # enumerate (index, value)
    print(index, item)
```

```python
# Hash : unique address of the object in memory
print(hash(1)) # hash of an integer
print(hash("alyan")) # hash of a string

t = ("apple", "banana")
print(hash(t))

# hash is used to check the immutability of the object, if the hash value of an object changes then it is mutable, if the hash value of an object remains the same then it is immutable
```

```python
# Nested Tuple 

nested_tuple = (1,2,(3,4), (5,6))
print(nested_tuple)
# Accessing elements in a nested tuple
print(nested_tuple[2]) # (3,4)
print(nested_tuple[2][0]) # 3
print(nested_tuple[3][1]) # 6

nst_tpl = (1,(2,3))
a , b = nst_tpl
print(a) # 1
print(b) # (2,3) # if only one variable is used to unpack the tuple then it will unpack the entire tuple into that variable

# but if we want to unpack the nested tuple into individual variables then we can use the following syntax
a , (b,c) = nst_tpl # now it will unpack each element of the nested tuple into individual variables
print(a) # 1
print(b) # 2
print(c) # 3


```
