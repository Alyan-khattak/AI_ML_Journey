### Private Attributes and Methods ###

    used only within class and can't be accessed outside class

    use " __ "" before attribute or methhod to make it private i.e __name, __display()

    syntax -->>  __attribute ,  __method()

IN pyhton private is not Trully private like Java but only Conceptually 

    exam:   class Person:
                def __init__(self, name):
                   self.__name = name  -->> name is private 

            p = Person("Alyan")
            print(p.__name)  -->> it will give error

        what python do is  restrict its use by this name (__name) but rather change the name of variable or method.

        __name changes to "_person__name" --> can be accessed using this 

        so "__attribute" becomes "_class__attribute"

## Getter and Setter ###

Use getter and Setter for private attributes



```python
## Del Keyword

class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print(f"Name : {self.name} , Age : {self.age}")

s1 = Student("Alyan", 21)
print(s1.display())
del s1
print(s1.display()) # now it will give error

```

```python
class Account:
    def __init__(self, acc_no, pin):
        self.acc_no = acc_no
        self.__pin = pin  ## this is private attribute "__pin"


    def display(self):
        print(f"Acc No : {self.acc_no} , PIN : {self.__pin} ")

    ## this is a private method
    def __show_pin(self):
        print("PIN : " , self.__pin)


a1 = Account(12344433,8683)
print(a1.acc_no)
#print(a1.__pin) # this won't get printed can only be used inside class
print(a1._Account__pin)  # cam accesss private varibale using _class__variable
print(a1.display())

# calling private method outside class
#print(a1.__show_pin()) # will give error
print(a1._Account__show_pin())
    
```

```python
## Using Getter and Setter
class GameCharater:
    def __init__(self, name, health):
        self.name = name
        self.__health = health # private

    #Getter
    def get_health(self):
        return self.__health

    #setter  ( in setter we write checks)
    def set_health(self, value):
        if value < 0:
            self.__health = 0
        elif value > 100:
            self.__health = 100
        else:
            self.__health = value

    def take_damage(self,damage):
        self.set_health(self.__health - damage)

    def heal(self, amount):
        self.set_health(self.__health + amount)

    

p1 = GameCharater("Alyan", 80)
print(p1.get_health())

p1.take_damage(55)
print(p1.get_health())

p1.heal(56)
print(p1.get_health())
```
