# Decorator #
A function that modifies another function without changing its code.
* in Function Closure in Return Statement we Could return executed Func i.e Func() and Func bith
        IN Decorator We must return the Function itseflf NOt Execute it

## Sytax ## 
    write with " @ " 
    
    def decorator(func):

        def wrapper():
                # extra behavior
                func()

        return wrapper

Explanation:
- decorator() receives a function as input.
- wrapper() adds extra behavior.
- wrapper() calls the original function.
- The decorator returns the wrapper function.

# Rule #
- Must Take a Function As Input
- Must Return another Function

```python
def after_message(func):

    def wrapper():
        func()
        print("function finished")
    return wrapper()

def work():
    print("working (passed using Closure)")

after_message(work)
```

```python
def call_counter(func):

    count = 0

    def wrapper():
        nonlocal count
        count += 1
        print("Function called Count", count,  "Times")
        func()

    return wrapper

@call_counter
def hello():
    print("hello")


hello() # how hello works .. bassiclly Call_counter is called and hello is passed as parameter
hello() # call_counter(hello)
hello()
```

```python
# Double the Result 

def double(func):

    def wrapper():
        rslt = func()
        return rslt * 2

    return wrapper

@double
def number():
    return 5 

print(number())
```

# Decoratoor With Arguements #
Some Times we don't want decorator to hard coded i.e
- Repeat a function N times (N can vary)
- Add custom messages or thresholds

                for these we need decorator that takes arguments

## Syntax ##
        def decorator_with_args(arg1, arg2):
                def actual_decorator(func):
                        def wrapper():
                        # use arg1 and arg2 here
                        func()
                        return wrapper
                return actual_decorator

- Outer function (decorator_with_args) — receives arguments for the decorator.

- Middle function (actual_decorator) — receives the function to be decorated.

- Inner function (wrapper) — runs the original function and adds extra behavior.

- Finally, decorator_with_args() returns the actual decorator, which Python applies to the function.

### Analogy ###
Decorator with arguments always has 3 layers:

- Outer function → receives decorator arguments (n)
- Middle function → receives the original function (func)
- Inner function → executes the original function with added behavior
        

```python
## reapeat a function n times

def repeat(n):

    def decorator(func):
        def wrapper():
            for _ in range(n):
                func()
        return wrapper
    return decorator

@repeat(3)
def hi():
    print("Hi there")

hi()  # @repeat(3) = hi = repeat(3)(hi)

# repeat(3) is called first and it return the decorator
# STEP 2: Apply the decorator to the function
# now decorator(hi) works 
# decorator returns wrapper 
# now hi → wrapper hi points to wrapper (as wrapper object is returned)

# Calling hi() python actually calls wrapper
```

```python
# Custom msg Before a function

def msg(msg):
    def decorator(func):
        def wrapper():
            print(msg)
            func()
        return wrapper
    return decorator

@msg("Alyan The Goat")
def greet():
    print("Hello")

greet()  # greet = msg(msg)(greet)
```

```python
# Two Arguements 
def reapeat_msg(n,msg):
    def decorator(func):
        def wrapper():
            for _ in range(n):
                print(msg)
                func()
        return wrapper
    return decorator

@reapeat_msg(4,"Alyan The Great")
def greet():
    print("hi")

greet()
```

```python
def before_msg(n, msg):  # decorator arguments
    def decorator(func):  # receives the function to wrap
        def wrapper():
            for _ in range(n):  # repeat n times
                print(msg)      # print the message
            func()             # call original function once
        return wrapper
    return decorator

@before_msg(3, "Starting now")
def task():
    print("Task running")

task()  # prints "Starting now" 3 times, then "Task running"

# Step 1: before_msg(3, "Starting now") is called → returns decorator
# Step 2: decorator(task) is called → returns wrapper
# Step 3: task now points to wrapper
# Step 4: task() actually calls wrapper()
# Step 5: wrapper() prints msg 3 times, then calls original task()
```

```python
# repeat Function and print Total runs
def repeat_with_count(n):
    def decorator(func):
        def wrapper():
            for i in range(1, n+1):
                print(f"Run {i}:")
                func()
        return wrapper
    return decorator

@repeat_with_count(4)
def greet():
    print("Hello!")

greet()

# Step 1: repeat_with_count(4) → returns decorator
# Step 2: decorator(greet) → returns wrapper
# Step 3: greet now points to wrapper
# Step 4: greet() actually calls wrapper()
# Step 5: wrapper() prints "Run i:" and calls original greet() 4 times
```

# Universal Decorator #
A normal decorator (or decorator with fixed arguments) only works for functions that take no arguments or a specific set of arguments.

        Use *args and **kwargs inside the wrapper. This makes the decorator universal, i.e., it can handle any function.

## syntax ##
        def universal_decorator(func):       # receives any function
                def wrapper(*args, **kwargs):    # wrapper can accept any number of arguments
                        # pre-processing (optional)
                        result = func(*args, **kwargs)  # call original function
                        # post-processing (optional)
                        return result
                return wrapper


| Feature                         | Fixed-Arg Decorator               | Universal Decorator                |
| ------------------------------- | --------------------------------- | ---------------------------------- |
| Accepts arguments for decorator | Yes (e.g., `@repeat(3)`)          | Optional (can also combine)        |
| Accepts any function signature  | No (works only for specific args) | Yes (`*args, **kwargs`)            |
| Complexity                      | Simple                            | Slightly more general, flexible    |
| Use case                        | Fixed logic                       | Real-world general-purpose wrapper |


```python
def log(func):                         # universal decorator
    def wrapper(*args, **kwargs):      # accepts any arguments  # wrapper will be like this when ur function has some arguments i.e add(5,7)
        print(f"Function {func.__name__} called with args={args}, kwargs={kwargs}")
        result = func(*args, **kwargs)
        print(f"Function {func.__name__} finished")
        return result
    return wrapper

@log
def add(a, b):
    return a + b

@log
def greet(name):
    print(f"Hello {name}")

print(add(5, 7))  # add = log(add)
# Decorator creates wrapper(*args, **kwargs). AND returns it
# so now add → wrapper , add points to wrapper 
# add(5,7) == wrapper(5,7) 
# inside wrapper : args collects all positional arguments into a tuple.
#                  **kwargs collects all keyword arguments into a dictionary.

# inside wrapper func is og add func we passed
# result = func(*args, **kwargs) -->> *args unpacks (5, 7) and passes them as: func(5, 7) 

## step By step Flow
# @log = add = log(add)
# Step 1: log(add) is called → wrapper(*args, **kwargs) is created
# Step 2: log(add) returns wrapper → add now points to wrapper
# Step 3: Calling add(5, 7) actually calls wrapper(5, 7)
# Step 4: wrapper receives arguments as args=(5,7), kwargs={}
# Step 5: wrapper calls original function: func(*args, **kwargs) → add(5,7)
# Step 6: wrapper prints pre-message, runs original function, prints post-message, returns result

greet("Alyan")
```

```python
# Upper case String Return 

def upper_case(func):
    def wrapper(*args, **kwargs):
        rslt = func(*args, **kwargs)
        return rslt.upper()
    return wrapper

@upper_case
def greet(name):
    return f"Hello {name}"

print(greet("Alyan"))
```

```python
# repeat N times

def repeat_n_times(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                print("Running Function")
                func(*args, **kwargs)
        return wrapper
    return decorator


@repeat_n_times(3)   # repeat_n_times(3) = repeat_n_times(3)(func)
def greet(name):
    print(f"Hello {name}")

greet("Alyan")

# epeat_n_times(3) → returns decorator

# decorator(greet) → returns wrapper

# greet = wrapper → greet("Alyan") calls wrapper

# wrapper(*args, **kwargs) → passes "Alyan" to original function
```

```python
# Conditional Execution 

def run_if(condition=True):
    def decorator(func):
        def wrapper(*args, **kwargs):
            if condition:
                return func(*args, **kwargs)
            else:
                print(f"{func.__name__} skipped")
        return wrapper
    return decorator

@run_if(True)
def greet(name):
    print(f"Hello {name}")

@run_if(False)
def bye(name):
    print(f"Bye {name}")

greet("Alyan")
bye("Ali")
```
