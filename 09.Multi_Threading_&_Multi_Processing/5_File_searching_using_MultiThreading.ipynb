# Parallel File Search System
- Multiple text files
- You want to search for a keyword (e.g., "ERROR")
- Do it faster using threads

```python
import threading
```

```python

# store results per thread
results = []

def search_file(filename, keyword):
    count = 0
    
    with open(filename, "r") as f:
        for line in f:
            if keyword in line:
                count += 1
    
    # Each thread appends its own result (safe enough here)
    results.append((filename, count))


# -------- MAIN --------

files = ["file1.txt", "file2.txt", "file3.txt"]
keyword = "ERROR"

threads = []

for file in files:
    t = threading.Thread(target=search_file, args=(file, keyword))
    threads.append(t)
    t.start()

for t in threads:
    t.join()

# combine results
total = sum(count for _, count in results)

print("Individual Results:", results)
print("Total occurrences:", total)
```

- Main Thread:
- ↓
- Creates 3 threads (one per file)
- ↓
- Thread-1 → searches file1
- Thread-2 → searches file2
- Thread-3 → searches file3
- ↓
- Each thread calculates its OWN count
- ↓
- Each thread appends result
- ↓
- Main thread waits (join)
- ↓
- Main thread combines results
