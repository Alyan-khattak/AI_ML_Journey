### Lambda FuncTion ###
Small , Anonymous (a function with a name ) defined using lambda keyword
* it can multiple arguements ( which you pass to it )
* It can have only one Expression ( one logic / one operation)

# i dont want to name this func , I just want its behavior #

# Syntax -->>  lambda arguements: expression   

```python
# Syntax
lambda arguements: expression 
```

```python
# Add function ( it has only one operation/logic but have to write all this)
def add(a,b):
    return a+b

add(3,4)
```

```python
# The above add function using lambda
addition = lambda a,b : a+b  #addition is variable that store lambda function
            #arguements : logic   (it will recieve a,b and perfoerm a+b)
print(addition(3,4))  #just call variable and give it parameters you want
```

```python
# functoion for Even 
def even(num):
    if num%2 == 0:
        return True
    
even(24)
```

```python
# Above func usnig Lambda

even = lambda num : num%2==0
print(even(24))
print(even(21))
```

```python
# sqaure Function
def square(x):
    return x * x

square(2)
```

```python
#Using Lambda

sq = lambda x: x *x
print(sq(2))
```

```python
# check if number is positive or negative
check = lambda n : "positive" if n >= 0 else "negative"
print(check(3))

# Length of strng 
length = lambda str : len(str)
print(length("alyan"))

# prod of 4 nmbrs 
prod = lambda a,b,c,d : a*b*c*d
print(prod(2,3,45,6))

# palindrome check 
palindrome = lambda str : str == str[::-1]
print(palindrome("aroosa"))
```
