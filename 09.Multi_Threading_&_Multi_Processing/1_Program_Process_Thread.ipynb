# Thread:

- unit of execution within a proccess

> A thread = smallest unit of execution inside a process

> Single Threaded Process:

- now when want to make rectangle so click on rectange -->> another thread 

[ Task 1 ] → [ Task 2 ] → [ Task 3 ]

> Multi Thread Process:

- Has multiple threads 

Thread 1 → Cooking
Thread 2 → Serving
Thread 3 → Billing


| Feature       | Process  | Thread                  |
| ------------- | -------- | ----------------------- |
| Memory        | Separate | Shared                  |
| Speed         | Slower   | Faster                  |
| Isolation     | High     | Low                     |
| Communication | Hard     | Easy                    |
| Crash impact  | Safe     | Can crash whole process |


> Multithreading   → Many workers in SAME process (shared memory)
> Multiprocessing  → Many processes (separate memory)
> Async Programming → One worker, but smart (non-blocking)
