# Iterator #

* Allows Efficient Looping and Memory Management

Its an Object:
- Remembers its state (where it is in a sequence)
- Returns one value at a time
- Knows when to stop

## Syntax ##
    iter() → creates iterator
    next() → gives next value

    iter(iterable) -->> iter([1,2,4])

### Purpose ###
To loop through data

To process large data efficiently

To avoid loading everything into memory

To control iteration manually

To create custom looping behavior

# Reverse Iterator #
* same thing just start from end and moves in backword direction

## syntax ##
    reversed(iterable) -->> reversed([1,2,3])
    next() -->> prints 3 then 2 then 1

# Custome Iterator #


```python
## Class that counts from 1-N

class MyCounter():

    def __init__(self, max_value):
        self.max_value = max_value #where to stop ( when we start from Zero)
        self.current = 0

    # iter method
    def __iter__(self):
        return self # returns Current Object
    
    #next method
    def __next__(self):
        if self.current < self.max_value: # if current is less than MAX
            self.current += 1
            return self.current # returns Value
        else:
            raise StopIteration("There are NO Element left In Iterator")
    

## Implementaion
c = MyCounter(10)

# __iter__() returns the c objct and Store it in " it "
it = iter(c)  #we make magic method so now we pass Objct instead of iterable
print(next(it))
print(next(it))
print(next(it))



```

```python
## Even Nmbr Iterator ##

class EvenNumber:
    def __init__(self, max_value):
        self.max_value = max_value
        self.current = 0

    def __iter__(self):
        return self
    
    def __next__(self):
        self.current += 2
        if self.current <= self.max_value:
            return self.current
        else:
            raise StopIteration("No elements left in Iterator")
        
## Implemenatation

e = EvenNumber(20)

it2 = iter(e)

while True:
    try:
        print(next(it2))
    except StopIteration as ex:
        print(ex)
        break
```

```python
# Fibonacii Iterator 

# Iterable Class 
class Fibonacci:

    def __init__(self, max_val):
        self.max_val = max_val

    def __iter__(self):
        return FibonacciIterator(self.max_val) # pass Max val to iterator, 
                                               # iterator objct is created and returned to it fib
        # here we called FibnaciiIterator(50),  it created an object of Fibancciiterator Class and we return that object to it_fib
        # Through that object we call next() an as it_fib has now initailzed till max 50, so next() works and return us values
## Iterator Class 
class FibonacciIterator:
    def __init__(self, max_val):
        self.max_val = max_val
        self.a = 0
        self.b = 1

    
    def __next__(self): # it gets an object, i.e we give it_fib
        if self.a > self.max_val:
            raise StopIteration
        
        value = self.a
        self.a, self.b = self.b, self.a + self.b # unpacking concept
        return value
    

## Implementation

fib = Fibonacci(50)
it_fib = iter(fib) # makes and iterator Object
# 🔹 Calls fib.__iter__()
# 🔹 __iter__ creates new FibonacciIterator object
# 🔹 Returns iterator object → assigned to it_fib
# Memory: it_fib.max_val = 50, it_fib.a=0, it_fib.b=1
# 🔹 Fibonacci class is now out of the loop; iterator object handles everything


while True:
    try:
        print(next(it_fib)) # we call next on it_fib which is an object of FibonaciIterator Clss
    except StopIteration:
        print("max value Reached")
        break
        
        
```

```python
# -----------------------------
# Iterable Class
# -----------------------------
class PrimeNumbers:
    def __init__(self, max_val):
        self.max_val = max_val
        # Store the maximum value for prime generation

    def __iter__(self):
        # 🔹 When iter(obj) is called, create a new iterator object
        return PrimeIterator(self.max_val)
        #  Factory (PrimeNumbers) produces a worker (PrimeIterator)
        # The worker generates primes

# -----------------------------
# Iterator Class
# -----------------------------
class PrimeIterator:
    def __init__(self, max_val):
        self.max_val = max_val
        self.current = 2  # Start checking for primes from 2

    def __next__(self):
        # 🔹 Keep searching for the next prime until max_val
        while self.current <= self.max_val:
            n = self.current
            self.current += 1  # Prepare for next call
            if self.is_prime(n):
                return n  # Return the prime number
        # 🔹 No more primes → stop iteration
        raise StopIteration

    def is_prime(self, num):
        # 🔹 Check if a number is prime
        if num < 2:
            return False
        for i in range(2, int(num**0.5)+1):
            if num % i == 0:
                return False
        return True

# -----------------------------
# Implementation
# -----------------------------
primes = PrimeNumbers(20)  
#  Factory object created
# Memory: primes.max_val = 20

it_primes = iter(primes)  
# 🔹 Calls primes.__iter__()
# 🔹 Returns a PrimeIterator object
# Memory: it_primes.current = 2, it_primes.max_val = 20

while True:
    try:
        print(next(it_primes))  
        # 🔹 Calls it_primes.__next__()
        # 🔹 Returns next prime number directly to print()
    except StopIteration:
        print("All primes up to 20 are generated!")
        break
```
