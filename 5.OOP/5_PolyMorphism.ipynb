## Method Over Loading ##

```python
class Calc:

    def add(self, a, b):
        return a + b
    
    # python doesnt allow this type of overLoading
    def add(self, a, b, c):
        return a + b + c


c1 = Calc()
c1.add(2,3)
c1.add(2,3,5)

```

```python
class Calculate:

    # use defalt values
    def add(self, a,b = 0,c = 0):
        if b == 0 and c == 0:
            return a
        elif b == 0:
            return a + c
        elif c == 0:
            return a + b
        else:
            return a+b+c
        
        # now this handles all the cases 
            

c1 = Calculate()
print(c1.add(1,2))
print(c1.add(4,0,7))
print(c1.add(37,56,232))

# OverLoading Works 
        

```

# Complex Numbers # 
1i + 3j  -- real + imgnry part

## when we add it ##
- 1i + 3j 
- 3i + 9j   
--------
- 4i + 12j 

real adds with real and img with img 

so we are Making a Complex nmbr class and Write our own logic for " + " operator


```python
class Complex:
    def __init__(self, real, img):
        self.real = real
        self.img  = img

    def display(self):
        print(f"{self.real}i + {self.img}j")

    ## Add method -->> not a magic method
    def add(self, other): #self is current objct, Other is passed objct
        newReal = self.real + other.real
        newImg = self.img + other.img
        return Complex(newReal, newImg)
    
    ### MAGIC METHOD 
    ##Add##
    def __add__(self, other):
        newReal = self.real + other.real
        newImg = self.img + other.img
        return Complex(newReal, newImg)
    
    ##Subtract##
    def __sub__(self, other):
        newReal = self.real - other.real
        newImg = self.img - other.img
        return Complex(newReal, newImg)


n1 = Complex(2,4)
n2 = Complex(4,9)

## Add method
n3 = n1.add(n2) # n1 + n2 will give error
n3.display()    # as we make add method but havent overload + operator

## NOw using Add magic Method
n4 = n1 + n2  # we Overloaded " + " using magic method
n4.display()
n5 = n3 + n2
n5.display()

n6 = n3 - n2
n6.display()



```

```python
class Wallet:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance
        self.transaction = [] # store transaction history

     # -----------------------------
    # USER FRIENDLY STRING  using print(objct) normally we write display() func
    # -----------------------------
    def __str__(self): 
        return f"Wallet of {self.owner} has Balance: {self.balance}"
    
    # -----------------------------
    # ADDITION OPERATOR (+)
    # -----------------------------
    def __add__(self, other):
        if isinstance(other, Wallet): # is other a wallet
            new_balance = self.balance + other.balance
            return Wallet(self.owner, new_balance) # creates new objct and store it in w3
        return NotImplemented
    
    # -----------------------------
    # SUBTRACTION OPERATOR (-)
    # -----------------------------
    def __sub__(self, amount):
        if isinstance(amount, (int, float)): # is amount int or float
            self.balance -= amount
            self.transaction.append(f"Spent {amount}")
            return self # modifies current object
        return NotImplemented
    
    # -----------------------------
    # GREATER THAN (>)
    # -----------------------------

    def __gt__(self, other):
        return self.balance > other.balance
    
    # -----------------------------
    # EQUALITY (==)
    # -----------------------------

    def __eq__(self, other):
        return self.balance == other.balance
    

w1 = Wallet("Alyan", 4700)
w2 = Wallet("Adil" , 7332)

print(w1)  # __str__
print(w2)


w3 = w1 + w2  # __add__
print(w3)

w1 - 6739   # __sub__
print(w1)

print(w1 > w2)  # __gt__
print(w1 == w2)

print(w1)
print(w2)
```
