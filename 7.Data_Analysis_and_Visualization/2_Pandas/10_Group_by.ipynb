# Aggregate Functions 
- min() , max() , sum() etc

## Multiple Aggregate Functions
### Using Dict:

    df.groupby().agg({"col name": "func" , "col name2": "func"})

- specific Fnction on specfic col

### using list:

    df.groupby().agg( ["func", "func2] )

- apply these function on ever col

### merge

    df.groupby.agg( {"col name": ["func1" , "func2" ]} )

```python
genre = movies.groupby("Genre")
genre
```

```python
genre.sum().head(1)  # apply sum on all the col
```

```python
# apply specific function on specifc col

genre.agg( 
    {
        "Runtime": "mean",
        "IMDB_Rating": "mean",
        "No_of_Votes": "sum",
        "Gross": "sum",
        "Metascore": "min"
    }
)
```

```python
genre.agg( ["max", "min"])
# now each col has min max col
```

```python
genre.agg( 
    {
        "Runtime": ["min", "max"],
        "IMDB_Rating": ["min", "max"],
        "No_of_Votes": ["sum", "mean"],
        "Gross" :    ["sum", "max", "min"]
    }
)
```

```python
import numpy as np

np.random.seed(42)

characters = [
    "Naruto","Sasuke","Sakura","Kakashi","Itachi","Hinata","Shikamaru",
    "Neji","Rock Lee","Gaara","Tsunade","Jiraiya","Orochimaru",
    "Minato","Madara","Obito","Pain","Konan","Killer Bee","Yamato"
]

villages = ["Leaf","Sand","Mist","Stone","Cloud"]

teams = ["Team 7","Team 8","Team 10","Akatsuki","Anbu"]

missions = ["D","C","B","A","S"]

df = pd.DataFrame({
    "Character": np.random.choice(characters,500),
    "Village": np.random.choice(villages,500),
    "Team": np.random.choice(teams,500),
    "MissionRank": np.random.choice(missions,500),
    "ChakraLevel": np.random.randint(50,1000,500),
    "BattlesWon": np.random.randint(0,100,500),
    "BattlesLost": np.random.randint(0,50,500),
    "JutsuKnown": np.random.randint(5,200,500),
    "Age": np.random.randint(12,60,500)
})

df.head(10)
```

```python
# Count ninjas per village.
df.groupby("Village").size()
```

```python
# Average chakra level per village.
df.groupby("Village")["ChakraLevel"].mean().sort_values( ascending = True)
```

```python
# Total battles won per team.
df.groupby("Team")["BattlesWon"].sum().sort_values( ascending = True)
```

```python
# Count characters per village and team.
df.groupby(["Village", "Team"]).size()
```

```python
# Average chakra per village and team.
df.groupby(["Village", "Team"])["ChakraLevel"].mean()
```

```python
df.groupby("Village").agg({
    "ChakraLevel":["mean","max","min"],
    "BattlesWon":"sum"
})
```

```python
df.groupby("Team").agg({
    "Age":["mean","max"],
    "JutsuKnown":"mean",
    "BattlesWon":"sum"
})
```

```python
# Village with most battles won.
df.groupby("Village")["BattlesWon"].sum().sort_values(ascending=False)
```

```python
# find Nmbr of movies starting with A for each group 

def starts_A(group):
    return group["Series_Title"].str.startswith('A').sum()

genre.apply(starts_A)
```

```python
def normal(group):
    group["norm_rating"] = group["IMDB_Rating"].min() / (group["IMDB_Rating"].max() - group["IMDB_Rating"].min())
    return group

genre.apply(normal)
```
