```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=1000, n_features=2 , n_classes= 2, n_clusters_per_class=1, n_redundant=0)
```

```python
pd.DataFrame(X)
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y= pd.DataFrame(X)[1] ,hue= y)

# The Datapomts are linearly Separable so we will use linear SVC
```

___
# Train test Split
___

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y  ,test_size=0.30, random_state= 42)
```

___
# Model Training
___

```python
from sklearn.svm import SVC
```

```python
model = SVC(kernel="linear") 
# as the Datapoints are linearly separable so we will use kernal as linear 
# # kernal is used to Dimension when data points are not clearl separabele
```

```python
model.fit(X_train, y_train)
```

```python
# prediction
y_pred = model.predict(X_test)
```

```python
print("Actual Value : ", y_test.round(2))
print("predicted Value : ", y_pred.round(2))
```

___
# Metrics
___

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

# Non Linear Dataset 
___

```python
X, y = make_classification(n_samples=1000, n_features=2 , n_classes= 2, n_clusters_per_class=2, n_redundant=0)
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y) 
# see the dataset is non linear so we will use difffernt kernals 
```

# Train Test SPlit 

```python
X_train, X_test, y_train, y_test = train_test_split(X, y  ,test_size=0.30, random_state= 42)
```

___
# Model Training at differnt Kernels
___

```python
from sklearn.svm import SVC
```

```python
model_simple = SVC(kernel="linear")

model_simple.fit(X_train, y_train)

# prediction 
y_pred = model_simple.predict(X_test)

# accuracy 

print(accuracy_score(y_test, y_pred))

# observe ... The Accuracy Droped from 98 to 93 at linear kernal as the DaTaset is not linear 
```

```python
model_2 = SVC(kernel="rbf")

model_2.fit(X_train, y_train)

# prediction 
y_pred = model_2.predict(X_test)

# accuracy 

print(accuracy_score(y_test, y_pred)) # at RBf kernel Accuracy improved 

```

```python
model_3 = SVC(kernel = "sigmoid")

model_3.fit(X_train, y_train)

# prediction 
y_pred = model_3.predict(X_test)

# accuracy 

print(accuracy_score(y_test, y_pred)) # at RBf kernel Accuracy improved 

```

```python
model_4 = SVC(kernel= "poly")

model_4.fit(X_train, y_train)

# prediction 
y_pred = model_4.predict(X_test)

# accuracy 

print(accuracy_score(y_test, y_pred)) # at RBf kernel Accuracy improved

```

```python
model_5 = SVC(kernel="sigmoid")

model_5.fit(X_train, y_train)

# prediction 
y_pred = model_5.predict(X_test)

# accuracy 

print(accuracy_score(y_test, y_pred)) # at RBf kernel Accuracy improved 


# RBF has the highest Accueacy  
```

    RBF gave the highest Accuracy so we will use it
