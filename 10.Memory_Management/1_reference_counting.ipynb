# Reference Counting 

> Reference counting = number of variables pointing to an object in memory

```python 
"""
Object exists in memory
↓
Python counts how many references (names) point to it
↓
If count = 0 → object is deleted (memory freed)


a = 10

Object: 10
Reference count = 1 (a → 10)

##########################################

a = 10
b = a

del a

b → 10

Reference count = 1

"""

import sys
sys.getrefcount(object)

Always shows +1 extra reference
(because function argument also references it)

````python
import sys
````

````python
a = [10]  # one reference to 10 

print( sys.getrefcount(a) ) # function arguement is also ref:  so ttl = 2
````

````python
a = [1,2,3]  # 1
print( sys.getrefcount(a))  # 2

b = a 
print( sys.getrefcount(a)) # 3

c = a 
print( sys.getrefcount(a) ) # 4

del b

print( sys.getrefcount(a) ) # 3

"""
[1,2,3]  ← a, b, c
Ref count = 3

del b

[1,2,3]  ← a, c
Ref count = 2

one extra is from function call

"""
````

````python
class A:
    pass

obj = A()  # 1
x = obj   # 2
y = obj   # 3

print( sys.getrefcount(obj))  # 4
````

````python
a = [1,2,3]
a = None  # before a was refering a list now to nothing so ref cout zero
 
print( sys.getrefcount(a) ) # zero ( but we would get some garbage value)
````
