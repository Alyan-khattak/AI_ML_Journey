# Function Copy #
- Function copy means creating another reference to a function, so you can call it by another name or modify its behavior (like wrapping).

        syntax -->> Function():                --> O.G Function
                    copy_func = Function       --> function copied to copy_func 
- ( notice we didn't use () i.e copy_func = Function() becasues `that means Calling a function)
- even when we delete the O.G function that Function is Callable by the Copy

## Functions in Python ##
In Python, functions are first-class objects, meaning:

- Functions can be assigned to variables.

- Can be passed as arguments.

- Can be copied to create another function reference.

# Closure #
- It’s a function inside another function.

- The inner function “remembers” variables    from the outer function.

                syntax: def Outer(x):
                                def inner(y):
                                        return x+y
                                return inner

```python
## Copy Function 

def Welcome():
    return "Welcome, This Function Gonna Get Copied"

print("\nO.G Function ")
print(Welcome())

print("\ncopying Function ")
copy_wel = Welcome
print(copy_wel())

print("\n cALLING O.G FUNCTION IN STRING TYPE VARIABLE ")
str = Welcome() # str is String Type Variable where Welcome Returned a string because we Used () which means CAll
print(str)

## Now deleteing the OG functtion 
print("\nDELETING OG Func")
del Welcome
#print(Welcome()) # will give error

print("\n Callig Copy After Deleting OG ")
print(copy_wel())

```

# Closure Function #
- It’s a function inside another function.

- The inner function “remembers” variables from the outer function.

```python
# Closure Function

def main_welcome(msg):
    def sub_welcome_method():
        print("Welcome To Inner Function")
        print("This Variable is Passed to Outer Function: ", msg)

    return sub_welcome_method() # i am returning the inner function

#Outer 

main_welcome("Alyan")
```

```python
# Define the outer function
def main_welcome(msg):
    # This is the inner function
    # It has access to variables in the outer function (like msg)
    def sub_welcome_method():
        print("Welcome To Inner Function")
        print("This Variable is Passed to Outer Function: ", msg)
    
    # Return the inner function itself (do NOT call it yet)
    return sub_welcome_method  # here i am not calling "()" so outer method will return function objct not output

# --------------------------------------------------
# Using the closure

print(main_welcome(5)) # <function main_welcome.<locals>.sub_welcome_method at 0x7f78b6391220>  -->> function objct not output
# Call the outer function, which returns the inner function
fn = main_welcome("Alyan")  # fn now holds sub_welcome_method

# Now call the inner function
fn()  # This executes sub_welcome_method and prints the messages

# --------------------------------------------------
# You can also do it in one line:
main_welcome("Alyan")()  # Calls outer, gets inner, and immediately executes it
```

```python
def make_adder(n):

    def adder(x):
        return x + n
    return adder # here i am not calling "()" so outer method will return function objct not output

print(make_adder(5)) # <function make_adder.<locals>.adder at 0x7f78b6307b60>  -->> function objct not output

add_5 = make_adder(5)  # store Function object in add_5 variable and call it using this variable
add_10 = make_adder(10)

print(add_5(3))
print(add_10(6))
```

# passing an Inbuilt Function as Arguements #
- In pyhton Functions are objects , So we can pass them As Arguements
- Change its name etc

```python
def Passing_Inbuilt_func(func):
    # funct will reciece inbuilt function print
    # so now func == print
    # we use print functionality using func i.e print("this") is now func("this")
    # as function are objects so just as we can objects name  we can change functions too
    def wrapper():
        print("Before Function")
        func("This is the passed 'print' Function")
    return wrapper()

Passing_Inbuilt_func(print) # passed Inbuilt Function print
```

```python
## pasing len() function

def proccess_len(func, data): # data is list

    def wrapper():
        print("Calculating lengh")
        rslt = func(data) # instead of len(data)
        print(f"Length : {rslt}")
    
    return wrapper()

n = [1,3,4,6,7,8]
proccess_len(len, n) # passed len
```

```python
def sum_wrapper(func , nums):
    def wrapper():
        print("Summing numbers...")
        result = func(nums)  # func = sum
        print(f"Sum: {result}")
    return wrapper()

numbers = [10, 20, 30, 40]
sum_wrapper(sum, numbers)
```

# Can Also Pass user Defined Functions #


```python
def greet(name): #user defined 
    print(f"Hello, {name}!")

def custom_wrapper(func, person):
    def wrapper():
        print("Greeting starts")
        func(person)  # call passed function
        print("Greeting ends")
    return wrapper()

custom_wrapper(greet, "Alice")
```
