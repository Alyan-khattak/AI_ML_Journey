# Methods #
## Object/Instance Methods ##
    A method that works on a specific object.
    If your method needs self, it’s an instance method.
* use self

## Static Method ##
    Shared by all the Objects
    No attribue is used like (self) or (cls)
    used when the method doesn't need to access class or object attributes ( only for logic)
* use @staticmethod decorator

## Class Method ##
    Belongs to Class 
    recieve class as an 1st implicit arguement def method(cls, parameters)

* use @classmethod decorator 





| Type            | Uses self? | Uses cls? | Access Instance Data | Access Class Data | When To Use              |
| --------------- | ---------- | --------- | -------------------- | ----------------- | ------------------------ |
| Instance Method | ✅ Yes      | ❌ No      | ✅ Yes                | ✅ Yes             | When working with object |
| Class Method    | ❌ No       | ✅ Yes     | ❌ No                 | ✅ Yes             | When working with class  |
| Static Method   | ❌ No       | ❌ No      | ❌ No                 | ❌ No              | Utility/helper logic     |


## Class Method ##


```python
class Person:

    name = "no name" # class attribute not object

    def change_name(self, name):
        self.name = name # a new instance attribute is created (name) its not the class one 

    # to change class name also 
    # access with class -->> class.name
    def change_class_name(self,name): # this is not a class method 
        Person.name = name  # 

    #another way to change not using class metho is 
    ## self__class__.name (self.__class__ ) means that i am accessing this objects (self) class (__class__)
    def change_class_name_way2(self, name):
        self.__class__.name = name # this will also change

p1 = Person()
p1.change_name("Alyan")
print(p1.name) # it will print Alyan

# but when I try to print Class name
print(Person.name) # prints -> No name

# as both are different attributes 
# so we can not change it using normal method using object 

## using Person.name to change ( wriiten in method )

p1.change_class_name("Adil") # now it will change
print(Person.name)
```

```python
## Class Methods ##
class Student:

    name = "No name"
    roll_no = "000"

    def __init__(self, name, roll_no):
        self.name = name
        self.roll_no = roll_no

    #Now using Class method to change above Class Variables
    @classmethod
    def Change_class_var(cls, name, roll_no): # cls is refrence to class just like self is to object
        cls.name = name
        cls.roll_no = roll_no 


s1 = Student("Alyan", "009")
print(s1.name)

print("printing Class var: ",Student.name)

s1.Change_class_var("Aroosa" , "083")
print("after Changing : ", Student.name)
```

```python
class Rectangle:

    def __init__(self, length, width):
        self.length = length
        self.width = width

    @property  # area will be calculted automatically and convrt to attribute
    def area(self):
        return self.length * self.width

rec1 = Rectangle(10, 5)

print("Area:", rec1.area)

# Change length
rec1.length = 20

print("After change:", rec1.area)
```
