# BroadCasting #
-Broadcasting is a mechanism that allows NumPy arrays of different shapes to perform element-wise operations by automatically expanding the smaller array to match the larger one.



- we see the smaller array repeats so that i could be Compatible with larger one

## Rules ##

-When two arrays interact, NumPy compares their shapes from right to left.

        A shape: (3,4)
        B shape: (4,)    --->> numpy interprets B as (1,4)
        then it sreches the one till 3 (3,4)  -->> now both have same shape

1: if nmbr of dimensions of two arrays are different, add new dimension with size 1, with head of array( left side ) with smaller array  ( like above we add 1) 

    Match the dimenstion for exm: A (3,3,3) , B (3,) 
    so add one B (1,1,3)  now Both array has same dimension


2: Make Each Dimension of Two Array the same Size
- Always strech 1
- if there is no 1, then it can't be strech and no broadcasting

        Above A(3,3,3,) , B(1,1,3)  -->> strech both ones to match Array A i.e till 3 
        so B (3,3,3)

- Example: A (1,3) , B(3,1)
    - Rule 1 alrdy Statisfied as Both Have same Dimenstion
    - Rule 2 : Strech 1 to make them same so A(1,3) strech 1 of A to match 3 of B(3,1) so A becomes A(3,3) 
    - now Strech 1 of B to match 3 of A(1,3) so B (3,3)

    - Now Both A and B becomes A(3,3) , B(3,3)






```python
import numpy as np
```

```python
a = np.array([1,2,3])
b = 5

print( a + b)
print()

a1 = np.arange(6).reshape(2,3)  # (2,3)
b1 = np.array([10,20,30])       # (,3) -->> (1,3) --->>(2,3)
print(a1+b1)
print()

a2 = np.arange(6).reshape(2,3)
b2 = np.array([[10],
               [20]])  # (2,1) -->> (2,3)
print(a2+b2)
print()

A = np.ones((3,4))
B = np.array([10,20,30,40])

print( A + B)
```

```python
a3 = np.arange(12).reshape(4,3)
b3 = np.array([1,2,3]) # (,3) -->> (1,3) -->>> (4,3)

print(a3 + b3 )
print()

```
