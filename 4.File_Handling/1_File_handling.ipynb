### File Handling ###
    Open the file
    Do something (read / write / append)
    Close the file

* ->> Text File == .txt ,.doc , .log
* -->> Binary File == .mp4, .img ,.jpg etc
### Modes ###
| Mode   | Meaning           |
| ------ | ----------------- |
| `"r"`  | read              |
| `"w"`  | write (overwrite) |    
| `"a"`  | append            |
| `"x"`  | create new file   |
| `"rb"` | read binary       |
| `"wb"` | write binary      |

## Methods #

* open("file_name", "mode")
* read()         -->> return file content in form of string
* close()  -->> close opened file
* readline() -->> reads one line at a time

* write("new data") -->> writes in file when opened in write mode





```python
f = open("demo.txt", "r")  #we oppen file in f 
data = f.read()    # f.read() will return file content in form of string
print(data)
f.close()
```

```python
f = open("demo.txt" , "r")
line1 = f.readline() # this read one lines at a time (this will also read one new line "\n" at the end of each line)
line2 = f.readline()
print(line1)
print(line2)
f.close()
```

```python
# writing a file 
f = open("demo.txt" , "w")
f.write("I am writing this using 'w' mode")
f.close()  # the above will now override in our file

f = open("demo.txt" ,"r")
data1 = f.read()  # the new thing will print
print(data1)  

f = open("demo.txt", "a")
f.write("now I am appending in the file")
f.close() # this line will append

f = open("demo.txt" , "r")
data2 = f.read()
print(data2)
f.close()
```

```python
# Opening in read+write Mode  use "r+" as mode

f  = open("demo.txt" , "r+")
f.write("I am using read+write mode now") # it will write at start of line 
print(f.read()) # as the above write wrote till now , so the pointer is at now -->> when we read it it will start from there
f.close()
```

### Open file using "with" Keyword """

Syantax --->> with open("file", "mode") as file:

    Open -->> Creates a file object
    as file (alias) -->> gives that object a name we can use and see
    with automatically closes the file so no need  to close it manualy
its indentation is same as loop or conditionals

with creates a safe temporary context where resources are automatically cleaned up when you leave.
    

```python
with open("sample.txt", "r") as file:
    content = file.read()
    print(content)
```

```python
# read file line by line

with open("sample.txt" , "r") as file:
    for line in file:
        print(line)  # line takes /n caharater also 

print("\nremoving the new line use .strip \n")

with open("sample.txt", "r") as f:
    for line in f:
        print(line.strip())  # .strip() removes \n (new line chaarcte)
```

```python
# Writing a file 
with open("example.txt" , "w") as file:
    file.write("this is the new file \n")
    file.write("this is new line")
```

```python
# Writing multiples lines without Overwriting -->> use "a"
lst = ["this is line1\n", "this is line2\n", "this is line3\n"]
with open("example.txt", "a") as file:
    file.writelines(lst)

with open("example.txt", "r") as file:
    content = file.read()
    print(content)
```

```python
# Read Content from source file and write it in destnation file
with open("sample.txt" , "r") as src_file:
    content = src_file.read()

with open("destination.txt" , "w") as dst_file:
    dst_file.write(content)

with open("destination.txt", "r") as file:
    content = file.read()
    print(content)
```

```python
# Count Lines Words and Characters in a file

with open("sample.txt", "r") as file:
    lines = 0
    words = 0
    chars = 0

    for line in file:
        lines += 1
        words += len(line.split())
        chars += len(line)
print("Lines", lines)
print("Words ", words)
print("characters ", chars)
```
