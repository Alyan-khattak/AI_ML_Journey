# Logging 
- Logging is a way to track what your program is doing automatically and cleanly.
- Tracks Events, errors & Oerational performance -->> keep logs of everything ( help in debugging )

## WHat Problem does it Solve

- Example problem:
- Your app crashes 😵

- Without logging:
    - 👉 "What happened??"

- With logging:
    - 👉 "Oh at 10:32, user input caused division by zero"


## Levels In Logging

- Logging levels are basically priorities / severity tags for messages.
        - “This is just info”
        - “This is dangerous”
        - “This is a serious error”

> Logging levels follow a strict order (hierarchy):

> DEBUG < INFO < WARNING < ERROR < CRITICAL

| Level    | Value |
| -------- | ----- |
| DEBUG    | 10    |
| INFO     | 20    |
| WARNING  | 30    |
| ERROR    | 40    |
| CRITICAL | 50    |

- we set Level in logging Basic Configuration

- if we set a lvl everything above will show ( like in config if we set *DEBUG* then all above this will show i.e can print msg through them )
- but if we set lvl *ERROR* only *CRITICAL* will show as it the only one above it, blow things would show

## Basic Syntaxx

```python
# simple
logging.basicConfig(
    level=logging.WARNING,
    
)

# advance 

import logging

logging.basicConfig(
    level=logging.DEBUG,
    format="%(asctime)s - %(name)s - %(levelname)s - %(message)s", # how the msg will display
    datefmt="%Y-%m-%d %H:%M:%S",  # Y=year, m=Month, d=Day, H = hr, M = mint, s = sec
    filename="app.log",  ## create app.log file and store logs there
    filemode="w"  # write mode  of file
)


```

| Placeholder     | Meaning             |
| --------------- | ------------------- |
| `%(asctime)s`   | Time of log         |
| `%(levelname)s` | Level (INFO, ERROR) |
| `%(message)s`   | Actual message      |
| `%(name)s`      | Logger name         |
| `%(filename)s`  | File name           |
| `%(lineno)d`    | Line number         |
