```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("birds_noisy_dataset.csv")
```

```python
missing_values = df.isnull().sum()
missing_percent = (missing_values / len(df)) * 100

missing_summary = pd.DataFrame({
    "Missing Count": missing_values,
    "Missing %": missing_percent
}).sort_values(by="Missing %", ascending=False)

print(missing_summary)
```

# stardarize Cols Names

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

# Smart way to Do 
- using Loop
    - convert to correct datatype
    - save memory 
    - fill miising values
    

```python
numeric_cols = [
    "wing_span_cm", "weight_g", "beak_length_mm", "tail_length_cm", 
    "age_years", "observation_year", "location_lat", "location_long", "temperature_c"
]

for col in numeric_cols:
    # Convert to numeric in case of dirty values
    df[col] = pd.to_numeric(df[col], errors="coerce")
    
    # Impossible values (negative or zero) → NaN
    df.loc[df[col] <= 0, col] = np.nan
    
    # Fill missing values with median (robust to outliers)
    df[col].fillna(df[col].median(), inplace=True)
    
    # Optimize memory
    df[col] = df[col].astype("float32")
    
    print(f"{col} cleaned ")
```

```python
df.info()
```

```python
# Catigorical Col

cat_cols = [col for col in df.columns if df[col].dtype == "str"]

for col in cat_cols:

    df[col] = df[col].astype("category")

    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].mode()[0] , inplace=True)
        

feature_cols = [col for col in df.columns if "feature_" in col]

# Feature Col
for col in feature_cols:
    df[col] = pd.to_numeric(df[col] , errors="coerce")
    df[col].fillna(df[col].mean(), inplace=True)
    df[col] = df[col].astype("float32")
```

```python
df.info()
```

```python
# Handling duplicate
print("Before duplicate removal:", df.shape)
df = df.drop_duplicates()
print("After duplicate removal:", df.shape)
```

# Handling Missing Vlues

```python
# Drop col where mssing values is more than 32 percent 

threshold = len(df) * 0.68 # keep col that have atleast 68% non missing valued
df = df.dropna(axis=1, thresh=threshold)
```

# Filling Missing Values


```python
# Categorical
df["species"] = df["species"].fillna(df["species"].mode()[0])
df ["family"] = df["family"].fillna(df["family"].mode()[0])
df['continent'] = df['continent'].fillna(df['continent'].mode()[0])
df['country'] = df['country'].fillna(df['country'].mode()[0])
df['habitat'] = df['habitat'].fillna(df['habitat'].mode()[0])
df['diet'] = df['diet'].fillna(df['diet'].mode()[0])
df['migratory'] = df['migratory'].fillna(df['migratory'].mode()[0])
df['gender'] = df['gender'].fillna(df['gender'].mode()[0])

# Numeric
# Fill missing with median wing span, or median per species
df["wing_span_cm"] = df.groupby("species")["wing_span_cm"].transform( lambda x: x.fillna(x.median()))
df["weight_g"] = df.groupby("species")["weight_g"].transform( lambda x : x.fillna(x.median()))
df['beak_length_mm'] = df.groupby('species')['beak_length_mm'].transform(lambda x: x.fillna(x.median()))
df['tail_length_cm'] = df.groupby('species')['tail_length_cm'].transform(lambda x: x.fillna(x.median()))
df['age_years'] = df['age_years'].fillna(df['age_years'].median())
df['observation_year'] = df['observation_year'].fillna(df['observation_year'].median())
df['location_lat'] = df['location_lat'].fillna(df['location_lat'].median())
df['location_long'] = df['location_long'].fillna(df['location_long'].median())
df['temperature_c'] = df['temperature_c'].fillna(df['temperature_c'].median())

# featuer col

feature_col = [ col for col in df.columns if "feature_" in col]
for col in feature_col:
    df[col] = df[col].fillna(df[col].mean())
```
