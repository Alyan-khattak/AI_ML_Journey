# important Data Frame Methods

- Value Counts:

        df.value_counts()  -->> returns frequency of each item in series But in DF returns frequency of each row

- sort_values:

        df.sort_value()

        - if we want to sort data using a specific col 

                df.sort_values("col_name")

- Rank

        df.rank() -->> dont work on DF

        rank item on base of one col ( rank on the base of score)

        df.rank(["score"]).rank()

- sort_index()

        df.sort_index()

- set index:

        df.set_index("col_name") -->> this col will act as index

- reset index:

        def.reset_index()

- Rename:

        df.rename(columns={"old_name" : "new_name"})

- N unique;

        DF.unique() -->> nbr of unique Value

- Missing Values:

        df.isnull()
        df.notnull()
        df.hasnull()

        movies["year"].hasnull()

        df.dropna()  -->> drop missing ( nan ) values ( if in Df there is one missing value in row the Whole row raw is dropped )

        df.dropna(how="all") -->> now only those rows will drop that which filed with missing values
 
        df.fillna(value_to_replace_nan_with)  -->>

- Drop Duplictes:

        df.drop_duplicates()  -->> drop Duplicate rows

        it has hidden parameter keep i.e which raw to keep 

        df.drop_duplicated(keep="first) -->> keep 1st raw and removes other duplictaes

- Apply

        df.apply()

```python
data = {
    "name": ["Alice", "Bob", "Charlie", "Alice", "Bob", "Eve", None],
    "score": [90, 85, 90, 90, 85, 70, np.nan],
    "year": [2021, 2020, 2021, 2021, 2020, 2019, 2022],
    "city": ["NY", "LA", "NY", "NY", "LA", None, "SF"]
}

df = pd.DataFrame(data)
print(df)
```

```python
print( df["name"].value_counts())
print()
print( df.value_counts())
```

```python
print( df.sort_values("year"))
print()
print(df.sort_values("year", ascending=False))
```
