```python
#Filter even numbers from list 

def filter_even(num):
    even = []
    for n in num:
        if n%2 == 0:
            even.append(n)
    return even

num = [i for i in range(1,20)]
print(filter_even(num))
```

```python
#  return min and max values from a tuple 

def find_min_max(num):
    return min(num), max(num)

tpl = (2,4,89,-78)
low , high = find_min_max(tpl)
print(f"min : {low} , max : {high}")
```

```python
# count Frequecy of elements 

def count_frequency(items):
    freq = {}
    for item in items:
        if item in freq:
            freq[item] += 1
        else:
            freq[item] = 1
    return freq


letters = ["a", "b", "a", "c", "b", "a"]
print(count_frequency(letters))
```

```python
# List of Tuples ( separate Passed Students)

def passed_stds(record):
    passed = []
    for name, marks in record:
        if marks >= 50:
            passed.append(name)
        
    return passed



students = [("alyan",20), ("adil" , 45), ("asad" , 78) , ("sana" , 100), ("aroosa" , 98)]
print(passed_stds(students))
```

```python
# Get KEys of A Students

def top_scoreres(scores):
    rslt = []
    for name,marks in scores.items():  #.items returns key,value pairs
        if marks>=80:
            rslt.append(name)
    return rslt


marks = {"Alyan": 20, "Adil": 60, "Sana": 90, "Aroosa": 96 , "Asad": 67}
print(top_scoreres(marks))
```

```python
# Split even and odd into dictionary ( maintain a list to store values againts keys even and odd)

def split_even_odd(nums):
    result = {"even": [], "odd": []}

    for n in nums:
        if n % 2 == 0:
            result["even"].append(n)
        else:
            result["odd"].append(n)

    return result
    

n = [i for i in range(1,21)]
print(split_even_odd(n))
```

```python
#separate TYpes using *args 

def separate_types(*args):
    nums = []
    strs = []

    for item in args:
        if isinstance(item,(int, float)):  # isinstance(object, type)  -->> “Is this object an instance of this type?”
            nums.append(item)
        elif isinstance(item, str):
            strs.append(item)
    
    return tuple(nums), strs

result = separate_types(1, "a", 2.5, "b", 3)
print(result)
```
