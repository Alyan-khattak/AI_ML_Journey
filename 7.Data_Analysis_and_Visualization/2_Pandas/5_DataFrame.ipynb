## loc and iloc

- same as series

| Method | How it works              | Think of it as |
| ------ | ------------------------- | -------------- |
| `loc`  | Uses **labels**           | Name lookup    |
| `iloc` | Uses **position numbers** | Index position |


    df.loc[label]

    dfiloc[position]

exam:

        df.iloc[0]  -->> will give row 0 ( all the Colmns in that row)

- Multiple Cols and Rows


        df.iloc[row, col] -->> iloc[0:3, 0:3]

exam:

        df.iloc[0:4, 0:2] -->> will give rows 0,1,2,3 and cols only 0,1

# loc and iloc
- loc[label]
- iloc[index]

```python
# using Dictionary 
std_dict = {
    "name": ["Alyan", "Aroosa", "Shayan"],
    "iq": [100,35,76],
    "marks": [53,76,34],
    "age" : [323,54,12]
}


std_df = pd.DataFrame(std_dict) # keys are taken as col name
print(  std_df )

std_df.set_index("name",inplace= True) # set col "name" as index
print()
print(std_df)
```

```python
# single Row using loc
print( std_df.loc["Alyan"]) # fetches That Row 
print( std_df.loc["Aroosa"])
```

```python
#using iloc
print( std_df.iloc[0])
print()
print( std_df.iloc[2])
```

```python
# muliple Rows -->> slicing
std_df.iloc[0:2]  # gives you rows 0,1 and every col in that
```

```python
std_df.iloc[0:3]
```

```python
std_df.iloc[0:3, 0:2] # rows 0,1,2 and col 0 and 1 not the third one
```

# Data Filtering #


```python
# find all ipl season winner 
ipl.head(2)
```

```python
mask = ipl["MatchNumber"] == 'Final'
print(type(mask))  
new_df =  ipl[mask] # row with true are kept and rows with false are discarded # now this will have only thoses macthes that are final
new_df[ ["Season", "WinningTeam"] ]
```

```python
#How many SuperOvers have Ocurred
ipl.head(3)
mask = ipl["SuperOver"] == 'Y'
new_df = ipl[mask]
new_df.shape  #14 is the nmbrs

# above can also be wriiten as 
ipl[ipl["SuperOver"] == 'Y' ].shape
```

```python
# how many Matches has CSK won in Kolkata

mask = ipl["City"] == "Kolkata"
new_df = ipl[ (mask) & ( "WinningTeam" == "Chennai Super Kings" )]
new_df
# (condition1) & (condition2)
ipl[(ipl["City"] == "Kolkata") & (ipl["WinningTeam"] == "Chennai Super Kings")].shape[0]
```

```python
# Toss Winner is Match Winner Percentage
ipl.head(2)
ipl[ (ipl["TossWinner"] == ipl["WinningTeam"])].shape

# percentage
ipl[ (ipl["TossWinner"] == ipl["WinningTeam"])].shape[0]/ipl.shape[0]
# 51%
```

```python
# Movies with rating higher than 8 and votes > 8000
movies.head(2)
movies[ (movies["imdb_rating"] > 8 ) & (movies["imdb_votes"] > 8000)]
```
