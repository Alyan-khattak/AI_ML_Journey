# broadcasting

```python
a6 = np.arange(3).reshape(1,3)  # ( 1, 3 ) -->> (1->2->3 , 3) -->> (3,3)
b6 = np.arange(3).reshape(3,1)  # ( 3, 1 ) -->> (3,  1->2->3) --->> (3,3)
#                                  increment the one and match its coresponding nmbrs 

print( a6* b6 )
```

```python
a5 = np.arange(12).reshape(3,4)  # (3,4)
b5 = np.arange(3) # (,3) -->> (1,3) -->> ( 3, 3 ) shapes didnt match so won't broadcast 
#                  we can only increament 1
print(a5)
print("\n", b5)

print( a5 + b5 )
```
