# 🏷️ Label Encoding

When working in Machine Learning, we often face categorical data.
One simple way to convert categories into numbers is:

> ## ✅ Label Encoding

---

## 💡 Core Idea

Assign a **unique integer** to each category.

| Category | Label Encoded |
|----------|--------------|
| Red | 0 |
| Blue | 1 |
| Green | 2 |

Simple. Fast. Each category gets its own number.

---

## 📌 Basic Example

Suppose:

| Color |
|-------|
| Red |
| Blue |
| Green |
| Red |

After Label Encoding:

| Color | Encoded |
|-------|---------|
| Red | 0 |
| Blue | 1 |
| Green | 2 |
| Red | 0 |

Same category → same number. Always.

---

## 🔢 How Numbers Are Assigned

By default (in scikit-learn):
- Numbers are assigned **alphabetically**

| Category | Alphabetical Order | Label |
|----------|-------------------|-------|
| Blue | 1st | 0 |
| Green | 2nd | 1 |
| Red | 3rd | 2 |

So:
- Blue → 0
- Green → 1
- Red → 2

---

## 🧪 Dry Run — Step By Step

Suppose dataset:

| Animal |
|--------|
| Cat |
| Dog |
| Bird |
| Dog |
| Cat |

### Step 1: Find Unique Categories

Unique animals (sorted alphabetically):
- Bird
- Cat
- Dog

### Step 2: Assign Labels

| Animal | Label |
|--------|-------|
| Bird | 0 |
| Cat | 1 |
| Dog | 2 |

### Step 3: Replace Original Values

| Animal | Encoded |
|--------|---------|
| Cat | 1 |
| Dog | 2 |
| Bird | 0 |
| Dog | 2 |
| Cat | 1 |

---

## 🐍 Python Code

### Using scikit-learn

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

data = ['Cat', 'Dog', 'Bird', 'Dog', 'Cat']

encoded = le.fit_transform(data)

print(encoded)
# Output: [1 2 0 2 1]

print(le.classes_)
# Output: ['Bird' 'Cat' 'Dog']
```

### Reverse: Decode Back to Original

```python
decoded = le.inverse_transform([1, 2, 0])

print(decoded)
# Output: ['Cat' 'Dog' 'Bird']
```

### Using Pandas (Manual Way)

```python
import pandas as pd

df = pd.DataFrame({'Animal': ['Cat', 'Dog', 'Bird', 'Dog', 'Cat']})

df['Animal_Encoded'] = df['Animal'].astype('category').cat.codes

print(df)
```

Output:

| Animal | Animal\_Encoded |
|--------|----------------|
| Cat | 1 |
| Dog | 2 |
| Bird | 0 |
| Dog | 2 |
| Cat | 1 |

## ✅ When To Use Label Encoding

### 1. Ordinal Categories (BEST USE CASE)

Meaning: **Categories have a natural order**

Examples:

| Education | Encoded |
|-----------|---------|
| Primary | 0 |
| Secondary | 1 |
| University | 2 |

Order is preserved. ✅ Perfect.

### 2. Tree-Based Models

Tree models like:
- Decision Tree
- Random Forest
- XGBoost
- LightGBM

These models **split by value**, so fake numeric order does NOT mislead them.

> ✅ Safe to use Label Encoding with tree-based models.

---

## ❌ When NOT To Use Label Encoding

### Linear / Distance-Based Models

Models like:
- Linear Regression
- Logistic Regression
- KNN
- SVM
- Neural Networks

These models **do math** on numbers.

Example:

| City | Label |
|------|-------|
| Lahore | 0 |
| Karachi | 1 |
| Islamabad | 2 |

Model may calculate:

(Lahore + Islamabad) / 2 = Karachi ???
(0 + 2) / 2 = 1 → Karachi


That makes no sense. 🚫

> ❌ Use One Hot Encoding instead for such models.

---

## ⚖️ Label Encoding vs One Hot Encoding

| Feature | Label Encoding | One Hot Encoding |
|---------|---------------|-----------------|
| New columns created | ❌ No | ✅ Yes (one per category) |
| Preserves order | ✅ Yes | ❌ No |
| Works for nominal data | ❌ Risky | ✅ Safe |
| Works for ordinal data | ✅ Best | ❌ Loses order |
| Memory usage | ✅ Low | ❌ Higher |
| Good for tree models | ✅ Yes | ✅ Yes |
| Good for linear models | ❌ No | ✅ Yes |

---

## 👍 Advantages

| # | Advantage |
|---|-----------|
| 1 | **Simple and fast** — one line of code |
| 2 | **No extra columns** — memory efficient |
| 3 | **Perfect for ordinal data** — preserves meaning |
| 4 | **Works well with tree-based models** |

---

## 👎 Disadvantages

| # | Disadvantage |
|---|-------------|
| 1 | **Fake numeric ordering** — misleads linear models |
| 2 | **Not safe for nominal categories** — cities, colors, names |
| 3 | **Model may learn wrong relationships** from assigned numbers |

---

## 🧠 Key Rule To Remember

> **If order exists → Label Encoding ✅**
> **If no order exists → One Hot Encoding ✅**

| Category Type | Example | Best Encoding |
|--------------|---------|--------------|
| Ordinal (has order) | Education, Size, Rating | Label Encoding |
| Nominal (no order) | Colors, Cities, Names | One Hot Encoding |

````python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# -------------------------
# DATA
# -------------------------
df = pd.DataFrame({
    "Color": ["Red", "Blue", "Green", "Red"]
})

print("ORIGINAL DATA:")
print(df)

# -------------------------
# LABEL ENCODER
# -------------------------
encoder = LabelEncoder()

# learn + encode
encoded = encoder.fit_transform(df["Color"])
df["Color_Encoded"] = encoded

# -------------------------
# OUTPUT
# -------------------------
print("\nENCODED DATA:")
print(df)

# -------------------------
# INTERNAL CLASSES
# -------------------------
print("\nCLASSES LEARNED:")
print(encoder.classes_)

# -------------------------
# DECODE BACK
# -------------------------
decoded = encoder.inverse_transform(df["Color_Encoded"])

print("\nDECODED VALUES:")
print(decoded)
````

# Multiple Cols # 

When you did OneHotEncoding:

    OneHotEncoder().fit_Transform(df)

👉 It automatically:

detects ALL categorical columns
encodes ALL of them
manages each column internally

So it behaves like:

“One smart machine for the entire dataset”


## 🚀 2. LabelEncoder = ONE column at a time tool

it is designed for 1D input i.e One Column at a time ( it can't detect and encode every column automatically like OneHotEncoder)

`when we Do:`

    LabelEncoder().fit(df["Color"])

> it only undersand one feature at a tym 
> if we have multiples Features ( cols ) we would do it for each one

````python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# -------------------------
# DATASET 
# -------------------------
df = pd.DataFrame({
    "Color": ["Red", "Blue", "Green", "Red", "Blue", "Green"],
    "City": ["Lahore", "Karachi", "Lahore", "Islamabad", "Karachi", "Multan"],
    "Gender": ["Male", "Female", "Female", "Male", "Male", "Female"],
    "Plan": ["Basic", "Premium", "Basic", "Gold", "Premium", "Basic"]
})

print("ORIGINAL DATA:")
print(df)

# -------------------------
# LABEL ENCODERS (one per column)
# -------------------------
le_color = LabelEncoder()
le_city = LabelEncoder()
le_gender = LabelEncoder()
le_plan = LabelEncoder()

# -------------------------
# FIT + TRANSFORM EACH COLUMN
# -------------------------
df["Color_encoded"] = le_color.fit_transform(df["Color"])
df["City_encoded"] = le_city.fit_transform(df["City"])
df["Gender_encoded"] = le_gender.fit_transform(df["Gender"])
df["Plan_encoded"] = le_plan.fit_transform(df["Plan"])

# -------------------------
# FINAL OUTPUT
# -------------------------
print("\nENCODED DATA:")
print(df)

# -------------------------
# SHOW MAPPINGS
# -------------------------
print("\nCOLOR CLASSES:", le_color.classes_)
print("CITY CLASSES:", le_city.classes_)
print("GENDER CLASSES:", le_gender.classes_)
print("PLAN CLASSES:", le_plan.classes_)

# -------------------------
# DECODE EXAMPLE (PLAN COLUMN)
# -------------------------
decoded_plan = le_plan.inverse_transform(df["Plan_encoded"])
print("\nDECODED PLAN COLUMN:")
print(decoded_plan)
````

# Real ML PipeLine

````python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# =========================================================
# 🧠 STEP 1: FULL DATASET (RAW REAL-WORLD STYLE DATA)
# =========================================================
# This is the original dataset containing:
# - Features (X): Gender, City, Plan, Device
# - Target (y): Churn (Yes/No)
# ---------------------------------------------------------

df = pd.DataFrame({
    "Gender": ["Male", "Female", "Female", "Male", "Male", "Female"],
    "City": ["Lahore", "Karachi", "Lahore", "Islamabad", "Karachi", "Multan"],
    "Plan": ["Basic", "Premium", "Basic", "Gold", "Premium", "Basic"],
    "Device": ["Mobile", "Laptop", "Mobile", "Tablet", "Mobile", "Watch"],
    "Churn": ["No", "Yes", "No", "Yes", "No", "Yes"]
})

print("ORIGINAL FULL DATA:\n", df)


# =========================================================
# 🧠 STEP 2: TRAIN / TEST SPLIT (SIMULATING REAL ML SCENARIO)
# =========================================================
# WHY SPLIT?
# - Train → model learns patterns
# - Test → model is evaluated on unseen data (future simulation)
# ---------------------------------------------------------

train = df.iloc[:4]   # first part used for training

# Test data contains unseen categories to simulate real world
test = pd.DataFrame({
    "Gender": ["Male", "Female"],
    "City": ["Lahore", "NewCity"],   # unseen category ❌
    "Plan": ["Basic", "VIP"],        # unseen category ❌
    "Device": ["Mobile", "VR"],      # unseen category ❌
    "Churn": ["No", "Yes"]
})

print("\nTRAIN DATA:\n", train)
print("\nTEST DATA:\n", test)


# =========================================================
# 🧠 STEP 3: SPLIT FEATURES (X) AND TARGET (y)
# =========================================================

# X = INPUT FEATURES (what model will learn from)
X_train = train.drop("Churn", axis=1) # drop Churn Col and store the remaining featiures in X_train dataframe
X_test = test.drop("Churn", axis=1)


# Y = Output or Target feature ,,  So we Store in a separe dataframe and then lanel encode it
y_train = train["Churn"]
y_test = test["Churn"]

# y = OUTPUT TARGET (what we want to predict)
# LabelEncoder converts Yes/No → 1/0
target_encoder = LabelEncoder()

y_train_encoded = target_encoder.fit_transform(y_train)
y_test_encoded = target_encoder.transform(y_test)

print("\nTARGET CLASSES (mapping):", target_encoder.classes_)
print("Encoded y_train:", y_train)


# =========================================================
# 🧠 STEP 4: ENCODE FEATURES (ONE HOT ENCODING)
# =========================================================
# WHY OneHotEncoder?
# - prevents fake ordering
# - each category becomes independent column
# ---------------------------------------------------------

encoder = OneHotEncoder(handle_unknown="ignore", sparse_output=False)

# FIT → learn categories ONLY from training data
# TRANSFORM → convert training data into numeric format
X_train_encoded = encoder.fit_transform(X_train)

# TRANSFORM test using SAME learned structure (NO re-learning)
X_test_encoded = encoder.transform(X_test)


# =========================================================
# 🧠 STEP 5: CONVERT NUMPY OUTPUT → READABLE DATAFRAME
# =========================================================

train_df = pd.DataFrame(
    X_train_encoded,
    columns=encoder.get_feature_names_out()
)

test_df = pd.DataFrame(
    X_test_encoded,
    columns=encoder.get_feature_names_out()
)

print("\nENCODED TRAIN FEATURES:\n", train_df)
print("\nENCODED TEST FEATURES:\n", test_df)


# =========================================================
# 🧠 STEP 6: WHAT HAPPENED INTERNALLY (IMPORTANT FLOW)
# =========================================================
"""
FLOW SUMMARY:

1. Raw Data
   → categorical + target columns

2. Split
   → train (learning)
   → test (evaluation)

3. X, y split
   → X = features
   → y = target

4. Label Encoding (y)
   → Yes/No → 1/0

5. OneHot Encoding (X)
   → categories → binary columns

6. Fit only on train
   → learns known categories

7. Transform test
   → applies same mapping
   → unseen categories → all zeros

IMPORTANT RULE:
NEVER fit on test data (prevents data leakage)
"""
````

````python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# =========================================================
# 🧠 STEP 1: RAW DATA (REAL-WORLD SIMULATION)
# =========================================================
# This dataset contains:
# - Features (X): Gender, City, Device, Membership
# - Target (y): Purchased (Yes/No)
# ---------------------------------------------------------

df = pd.DataFrame({
    "Gender": ["Male", "Female", "Female", "Male", "Female", "Male"],
    "City": ["Lahore", "Karachi", "Lahore", "Islamabad", "Multan", "Karachi"],
    "Device": ["Mobile", "Laptop", "Mobile", "Tablet", "Watch", "Mobile"],
    "Membership": ["Basic", "Premium", "Basic", "Gold", "Premium", "Basic"],
    "Purchased": ["No", "Yes", "No", "Yes", "Yes", "No"]
})

print("RAW DATA:\n", df)


# =========================================================
# 🧠 STEP 2: TRAIN / TEST SPLIT (SIMULATING REAL WORLD)
# =========================================================
# WHY SPLIT?
# - TRAIN → model learns patterns
# - TEST → model evaluates on unseen future-like data
# ---------------------------------------------------------

train = df.iloc[:4]

# Test data includes unseen categories to simulate real-world scenarios
test = pd.DataFrame({
    "Gender": ["Male", "Female"],
    "City": ["Lahore", "NewCity"],     # unseen ❌
    "Device": ["Mobile", "VR"],        # unseen ❌
    "Membership": ["Basic", "VIP"],    # unseen ❌
    "Purchased": ["No", "Yes"]
})

print("\nTRAIN DATA:\n", train)
print("\nTEST DATA:\n", test)


# =========================================================
# 🧠 STEP 3: SPLIT FEATURES (X) AND TARGET (y)
# =========================================================
# X = INPUT FEATURES (what model learns from)
# y = TARGET LABEL (what we want to predict)
# ---------------------------------------------------------

X_train = train.drop("Purchased", axis=1) # include everything except Purchased Col
X_test = test.drop("Purchased", axis=1)

y_train = train["Purchased"]
y_test = test["Purchased"]


# =========================================================
# 🧠 STEP 4: ENCODE TARGET (y) USING LABEL ENCODER
# =========================================================
# WHY LABEL ENCODER?
# - ML models understand only numbers
# - Convert Yes/No → 1/0
# ---------------------------------------------------------

label_encoder = LabelEncoder()

y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

print("\nTARGET MAPPING (classes):", label_encoder.classes_)
print("Encoded y_train:", y_train_encoded)


# =========================================================
# 🧠 STEP 5: ENCODE FEATURES (X) USING ONE HOT ENCODER
# =========================================================
# WHY ONE HOT ENCODING?
# - avoids fake numeric ordering
# - creates independent binary columns
# - safe for categorical features
# ---------------------------------------------------------

encoder = OneHotEncoder(handle_unknown="ignore", sparse_output=False)

# FIT → learn categories ONLY from training data
# TRANSFORM → convert training data into numeric format
X_train_encoded = encoder.fit_transform(X_train)

# TRANSFORM test using SAME learned mapping
# unseen categories → automatically ignored → all zeros
X_test_encoded = encoder.transform(X_test)


# =========================================================
# 🧠 STEP 6: CONVERT ENCODED OUTPUT TO READABLE DATAFRAMES
# =========================================================
# WHY?
# - model works with arrays
# - humans need readable format for debugging
# ---------------------------------------------------------

train_encoded_df = pd.DataFrame(
    X_train_encoded,
    columns=encoder.get_feature_names_out()
)

test_encoded_df = pd.DataFrame(
    X_test_encoded,
    columns=encoder.get_feature_names_out()
)

print("\nENCODED TRAIN FEATURES:\n", train_encoded_df)
print("\nENCODED TEST FEATURES:\n", test_encoded_df)


# =========================================================
# 🧠 STEP 7: WHAT HAPPENED INTERNALLY (FINAL FLOW)
# =========================================================
"""
FULL FLOW SUMMARY:

1. RAW DATA
   → categorical + target columns

2. SPLIT DATA
   → train = learning phase
   → test = evaluation phase

3. X / y SPLIT
   → X = features (inputs)
   → y = target (output)

4. LABEL ENCODING (y)
   → Yes/No → 1/0

5. ONE HOT ENCODING (X)
   → categories → binary columns

6. FIT ONLY ON TRAIN
   → learns known categories

7. TRANSFORM TEST
   → applies same mapping
   → unseen categories → all zeros

IMPORTANT RULE:
NEVER fit on test data → prevents data leakage
"""
````

````python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# =========================================================
# 🧠 STEP 1: RAW REAL-WORLD DATASET
# =========================================================
# This simulates an e-commerce platform user behavior dataset
# ---------------------------------------------------------

df = pd.DataFrame({
    "Gender": ["Male", "Female", "Female", "Male", "Male", "Female"],
    "City": ["Lahore", "Karachi", "Lahore", "Islamabad", "Multan", "Karachi"],
    "Device": ["Mobile", "Laptop", "Mobile", "Tablet", "Watch", "Mobile"],
    "Category": ["Electronics", "Clothing", "Electronics", "Beauty", "Clothing", "Beauty"],
    "Membership": ["Basic", "Premium", "Basic", "Gold", "Premium", "Basic"],
    "Purchased": ["No", "Yes", "No", "Yes", "Yes", "No"]
})

print("RAW DATA:\n", df)


# =========================================================
# 🧠 STEP 2: TRAIN / TEST SPLIT (SIMULATING REAL WORLD)
# =========================================================

train = df.iloc[:4]

test = pd.DataFrame({
    "Gender": ["Male", "Female"],
    "City": ["Lahore", "NewCity"],         # unseen ❌
    "Device": ["Mobile", "VR"],            # unseen ❌
    "Category": ["Electronics", "Sports"], # unseen ❌
    "Membership": ["Basic", "VIP"],        # unseen ❌
    "Purchased": ["No", "Yes"]
})

print("\nTRAIN DATA:\n", train)
print("\nTEST DATA:\n", test)


# =========================================================
# 🧠 STEP 3: SPLIT FEATURES (X) AND TARGET (y)
# =========================================================

X_train = train.drop("Purchased", axis=1)
X_test = test.drop("Purchased", axis=1)

y_train = train["Purchased"]
y_test = test["Purchased"]


# =========================================================
# 🧠 STEP 4: ENCODE TARGET (LABEL ENCODING)
# =========================================================
# Convert Yes/No → 1/0 for ML model
# ---------------------------------------------------------

label_encoder = LabelEncoder()

y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

print("\nTARGET CLASSES:", label_encoder.classes_)
print("Encoded y_train:", y_train_encoded)


# =========================================================
# 🧠 STEP 5: ENCODE FEATURES (ONE HOT ENCODING)
# =========================================================
# WHY OneHot?
# - avoids fake ordering
# - safe for categorical variables
# - creates independent feature space
# ---------------------------------------------------------

encoder = OneHotEncoder(handle_unknown="ignore", sparse_output=False)

# LEARN ONLY FROM TRAIN DATA
X_train_encoded = encoder.fit_transform(X_train)

# APPLY SAME LEARNING TO TEST DATA
# unseen categories → automatically become all zeros
X_test_encoded = encoder.transform(X_test)


# =========================================================
# 🧠 STEP 6: CONVERT BACK TO READABLE DATAFRAMES
# =========================================================

train_df = pd.DataFrame(
    X_train_encoded,
    columns=encoder.get_feature_names_out()
)

test_df = pd.DataFrame(
    X_test_encoded,
    columns=encoder.get_feature_names_out()
)

print("\nENCODED TRAIN FEATURES:\n", train_df)
print("\nENCODED TEST FEATURES:\n", test_df)


# =========================================================
# 🧠 STEP 7: WHAT THIS PIPELINE ACTUALLY DID
# =========================================================
"""
FULL FLOW:

1. RAW DATA
   → multiple categorical features + target

2. SPLIT
   → train = learning dataset
   → test = unseen future-like data

3. X / y SPLIT
   → X = input features
   → y = output label (Purchased)

4. LABEL ENCODING (y)
   → Yes/No → 1/0

5. ONE HOT ENCODING (X)
   → each category becomes binary columns

6. FIT ONLY ON TRAIN
   → learns known categories

7. TRANSFORM TEST
   → applies same mapping
   → unseen categories → all zeros

KEY RULE:
NEVER fit on test data → avoids data leakage
"""
````
