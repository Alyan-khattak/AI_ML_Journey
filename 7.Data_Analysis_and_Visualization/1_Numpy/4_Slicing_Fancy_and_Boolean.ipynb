# Fancy Indexing #
- Some times we can't Pattern (like skip 1,2,3 .. rows Cols whch can be done using Steps)
- so we use Fancy Indexing : Here we can Tell which specific Row, Col we want even with no pattern like row 1,5,6,8,17
- for ID it can be said for Elements 

### 1D 
        array[[index1, index2, index3]]
array[ [] ] -->> gives list [] inside

### 2D 
        array[[rows]]   -->> in col not given then it takes rows
        array[[row_indexes], [col_indexes]]


## Boolean Indexing ##
- Boolean indexing selects elements based on conditions.

                > < >= <= == !=
                syantax: array[condition]

## Multiple Coditions 
- &  → AND
- |  → OR
- ~  → NOT

        array[ (condition 1 ) & (condition 2 )]

# Fancy indexing

```python
import numpy as np
```

```python
a = np.random.randint(1,20, 10) # ranint(start range, end range, nmber of random items)
print(a)
```

```python
print( a[ [ 0,3,6,8,9] ]) # only prints these indexes 
print( a[ [1,2,9,0] ] )
print( a[[8,0,1,2,3] ] )
print( a[ [ 1,2,3 ] ])
```

```python
# 2D
a1 = np.arange(24).reshape(6,4)
print(a1)
```

```python
# extract 1,2,,5,6 rows
print( a1[ [0,1,4,5 ] ])
print()
print(a1[[1,2,5]])
print()

# Extract 0,2,3 cols
print( a1[:,[0,2,3]])
print()
print( a1[:,[1,2]])
print()

print(a1[[0,1,2],[2,1,0]]) #print specific element giving row and col index
print()  # row 0, col2 : row 1 col 1 : row 2 col 0

print(a1[[4,2],[0,1]]) #16 and 9
print()

print( a1[[1,2,5]])
print()
print(a1[:,[1,3]])
```

# Boolean Indexing #

```python
a2 = np.random.randint(1,20,8)
print(a2)
```

```python
print(a[a>5]) # print elemenst greater than 5
print(a[a==19])
print( a[a<=8] )
print(a[(a>5) & (a%2 == 0)])
print( a[ (a<25) | (a>2)])
```

```python
# 2D
a1 = np.arange(24).reshape(6,4)
print(a1)
```

```python
print(a1[a1<9])
print(a1[ (a1>= 5 ) & ( a1 <=10 )])
print(a1 [ (a1 >= 6 ) & (a1 % 2 != 0)])
```
