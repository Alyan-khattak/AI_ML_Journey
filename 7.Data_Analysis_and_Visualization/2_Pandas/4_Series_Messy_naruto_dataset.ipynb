```python
import pandas as pd
import numpy as np
```

```python
# Load file
data = pd.read_csv("/home/aizen/AI_ML/7.Data_Analysis_and_Visualization/2_Pandas/messy_naruto_ai_dataset.csv")
print(data)
```

```python
# Extract Series 
names = data["Name"]
village = data["Village"]
rank = data["Rank"]
clan = data["Clan"]
power = data["Power_Level"]

type(power)
```

```python
power.dtype
print(power) # incorect Values " ?? "
print()
# Convert to numeric
power = pd.to_numeric(power, errors= "coerce") # "??" → NaN
print(power)   # errors="coerce" → invalid values become 
```

```python
power.dtype # now its float
```

```python
# Detect missing Values
print ( power.isnull().sum() )
# isnull() → returns True for missing values
# sum() → counts them

# replace them 
power = power.fillna(power.mean())
# fillna() → replaces missing values
# power.mean() → average value
print(power)
```

```python
# remove bad Values ( not between 0-100)

power = power.where((power >= 70) & (power <= 100) )
print(power)

# where(condition)
# keeps values if condition is true
# otherwise → NaN
```

```python
# replace nan
power = power.fillna(power.mean())
print(power)
```
