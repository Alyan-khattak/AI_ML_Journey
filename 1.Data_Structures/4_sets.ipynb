```python
# set are unque, unoreded and no duplicates
# perform mathematical operations like union, intersection, difference and symmetric difference

my_set = {1,2,3,4,4,5,}
print(my_set)

#type casting
my_list = [1,2,3]
print(type(my_list))
s = set(my_list)
print(type(s))
```

```python
#Basic Set Operations  Adding elements to a set
A = {1,2,3,4}
A.add(5)
print(A)

# removing elements from a set
A.remove(3)
print(A) 
A.remove(10) # this will raise an error because 10 is not in the set
print(A)

A.discard(10) # this will not raise an error because discard does not raise an error if the element is not found
print(A)
```

```python
#pop method
s = {1,2,3,4,5,6}
poped_element = s.pop()  # pop removes 1st ele from the set, whatever ele is 1st in the set, it will remove that ele and return it
print(poped_element)
print("reiming the set after poping an element", s )

poped_2nd = s.pop()
print(poped_2nd)
print("reiming the set after poping 2nd element", s )
```

```python
#Set membership Test : wheather an element is present in the set or not
my_sset = {1,2,3,4,5}
print(3 in my_sset) # True
print(10 in my_sset) # False

print(4 in my_sset)
```

```python
#mathematical operations on sets
set1 = {1,2,3,4}
set2 = {3,4,5,6}

#union 
union_set = set1.union(set2)
print("Union of set1 and set2 is:", union_set)

#intersection ( common elements in both sets)
intersection_set = set1.intersection(set2)
print("Intersection of set1 and set2 is:", intersection_set)    

# instersection update ( it will update the rslt of intersection in set 1 and set 1 will be updated with the common elements)
set1.intersection_update(set2)
print("set1 after intersection update with set2 is:", set1)
```

```python
s1 = {1,2,3,4}
s2 = {3,4,5,6}

# difference ( elements that are in set1 but not in set2)
difference_set = s1.difference(s2)
print("Difference of s1 and s2 is:", difference_set)

# differece upadate
s1.difference_update(s2)
print("s1 after difference update with s2 is:", s1)

#symmetric difference ( elements that are in either set1 or set2 but not in both) combine the unique elements from both sets
ss1 = {1,2,3,4}
ss2 = {3,4,5,6}
sym_diff_set = ss1.symmetric_difference(ss2)
print("Symmetric difference of ss1 and ss2 is:", sym_diff_set)
```

```python
#set Methods

a1 = {1,2,3,4}
a2 = {3,4,5,6}

print(a1.issubset(a2)) # False because all elements of a1 are not in a2
print(a2.issubset(a1)) # False because all elements of a2 are

print(a1.issuperset(a2)) # False because a1 does not contain all elements of a2
print(a2.issuperset(a1)) # False because a2 does not contain all elements of a1

print(a1.isdisjoint(a2)) # False because a1 and a2 have common elements (3 and 4)
```
