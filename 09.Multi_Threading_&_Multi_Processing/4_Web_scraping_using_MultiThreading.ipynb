#  Web Scraping

- we will practically implement MultiThreading in Web scraping 
- We will use Libraries like *requests* for HTTPS responses and Requests & *bs4* -->> *beautifulSoup* for scrapping 

```python
import threading
import requests
from bs4 import BeautifulSoup
```

```python
# Basic scrapper 

urls = [
    "https://www.python.org",
    "https://www.github.com",
    "https://www.stackoverflow.com",
    "https://www.wikipedia.org"
]

def fetch(url):
    print(f"fechting {url}")

    response = requests.get(url)
    print(f"{url} Status: {response.status_code}")

thread = []

for url in urls:
    t = threading.Thread(target=fetch, args=(url, ))
    thread.append(t)
    t.start()


for t in thread:
    t.join()
    
```

```python
# Extract page title using Beatiful Soup

urls = [
    "https://www.python.org",
    "https://www.github.com",
    "https://www.stackoverflow.com",
    "https://www.wikipedia.org"
]


def fetch_title(url):
    print(f"fetch {url}")

    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    title = soup.title.string if soup.title else "No title"
    print(f"{url} Title: {title}")


thread = []

for url in urls:
    t = threading.Thread(target=fetch_title, args=(url, ))
    thread.append(t)
    t.start()

for t in thread:
    t.join()
```

```python
import threading
import requests
from bs4 import BeautifulSoup

# store results (fixed size, no conflict)
results = []

def crawl(index, url):
    print(f"Fetching: {url}")
    
    try:
        response = requests.get(url, timeout=5)
        soup = BeautifulSoup(response.text, "html.parser")
        
        # extract links
        links = []
        for tag in soup.find_all("a", href=True):
            links.append(tag["href"])
        
        # store result at fixed index
        results[index] = {
            "url": url,
            "num_links": len(links)
        }
    
    except Exception as e:
        results[index] = {
            "url": url,
            "error": str(e)
        }


# -------- MAIN --------

urls = [
    "https://www.python.org",
    "https://www.github.com",
    "https://www.wikipedia.org",
    "https://www.stackoverflow.com"
]

# pre-allocate results (IMPORTANT)
results = [None] * len(urls)

threads = []

for i, url in enumerate(urls):
    t = threading.Thread(target=crawl, args=(i, url))
    threads.append(t)
    t.start()

for t in threads:
    t.join()

# -------- RESULTS --------

for result in results:
    print(result)
```
