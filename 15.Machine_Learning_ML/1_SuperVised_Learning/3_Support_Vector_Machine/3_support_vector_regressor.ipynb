# Support Vector Regression (SVR)

Support Vector Regression (SVR) is the **regression version of Support Vector Machine (SVM)**.

Instead of classifying data into classes, SVR predicts **continuous numerical values**.

Examples:

* House Price Prediction
* Stock Price Prediction
* Temperature Prediction
* Sales Forecasting

---

# Linear Regression vs SVR

### Linear Regression

Linear Regression tries to minimize the error for **every data point**.

```text
            *
         *
      *
   *
-----------------------------> x
```

Every point contributes to the loss.

---

### Support Vector Regression

SVR creates a **margin (tube)** around the regression line.

Only the points **outside the tube** affect the model.

```text
            *
====================== Upper Margin (ε)

------------- Best Fit Line -------------

====================== Lower Margin (ε)

      *
```

Points inside the margin are considered "good enough."

---

# Intuition

Imagine your teacher says:

> "If your answer is within **±2 marks**, I'll consider it correct."

Suppose the actual marks are:

```text
90
```

Predictions:

```text
89 ✅
91 ✅
92 ✅
88 ✅
```

All of these are acceptable because they are within **±2**.

SVR works the same way using an **epsilon (ε) margin**.

---

# What is Epsilon (ε)?

Epsilon defines the width of the margin (tube).

```text
          Upper Margin

===============================

----------- Regression Line -----------

===============================

          Lower Margin
```

Predictions inside this tube have **zero loss**.

Only predictions outside the tube are penalized.

---

# Why Use SVR?

Linear Regression:

* Tries to fit every point.
* Sensitive to noise and outliers.

SVR:

* Ignores small errors.
* Focuses only on important points.
* Produces a more robust model.

---

# Support Vectors

The points that lie **outside or on the edge of the epsilon margin** are called **Support Vectors**.

These points determine the final regression line.

```text
           *

====================

--------------------

====================

       *
```

The starred points are the support vectors.

---

# Types of SVR Kernels

Just like SVM, SVR supports different kernels.

| Kernel  | Use Case                              |
| ------- | ------------------------------------- |
| Linear  | Linear relationship                   |
| Poly    | Polynomial relationship               |
| RBF     | Non-linear relationship (Most Common) |
| Sigmoid | Rarely used                           |

# Important Hyperparameters

## 1. kernel

Specifies the kernel type.

```python
kernel="rbf"
```

Possible values:

```python
kernel="linear"
kernel="poly"
kernel="rbf"
kernel="sigmoid"
```

---

## 2. C

Regularization parameter.

```python
C=1.0
```

Controls the trade-off between:

* Keeping the regression line smooth.
* Fitting the training data closely.

### Small C

```text
More Regularization

Simpler Model
```

### Large C

```text
Less Regularization

Fits Training Data More Closely
```

---

## 3. epsilon (ε)

Controls the width of the margin.

```python
epsilon=0.1
```

### Small ε

```text
Narrow Tube

More points become support vectors.
```

### Large ε

```text
Wide Tube

More prediction errors are ignored.
```

---

## 4. gamma

Used with RBF, Polynomial, and Sigmoid kernels.

```python
gamma="scale"
```

Controls how much influence each training sample has.

### Small Gamma

```text
Smooth Decision Function
```

### Large Gamma

```text
Complex Decision Function

May Overfit
```

---

from sklearn.svm import SVR

model = SVR(

# Linear Regression vs SVR

| Feature                 | Linear Regression | SVR                |
| ----------------------- | ----------------- | ------------------ |
| Task                    | Regression        | Regression         |
| Fits Every Point        | Yes               | No                 |
| Uses Margin             | No                | Yes                |
| Robust to Small Errors  | No                | Yes                |
| Supports Kernels        | No                | Yes                |
| Handles Non-linear Data | No                | Yes (with kernels) |

---

# When Should We Use SVR?

Use SVR when:

* The dataset is small to medium-sized.
* The relationship is non-linear.
* You want robustness against small prediction errors.
* You need kernel methods to model complex patterns.

Examples:

* House price prediction
* Electricity demand forecasting
* Stock price prediction
* Weather forecasting
* Sales prediction

---

# Rule of Thumb

* **Linear Regression** → Simple linear relationship.
* **Polynomial Regression** → Curved relationship with explicit polynomial features.
* **SVR (Linear Kernel)** → Linear regression with a margin and regularization.
* **SVR (RBF Kernel)** → Complex non-linear regression problems.

# X and Y
___

````python
X = df.drop("total_bill", axis=1)
y= df["total_bill"]
````

___
# Train Test Split
___

````python
from sklearn.model_selection import train_test_split
````

````python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=42)
````

````python
X_train.head()
````

___
# Feature Encoding
___

````python
from sklearn.preprocessing import LabelEncoder
````

````python
le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()

# for Train Data

X_train["sex"] = le1.fit_transform(X_train["sex"])
X_train["smoker"] = le2.fit_transform(X_train["smoker"])
X_train['time'] = le3.fit_transform(X_train["time"])
````

````python
X_train.head()
````

````python
# for test Data 
X_test["sex"] = le1.transform(X_test["sex"])
X_test["smoker"] = le2.transform(X_test["smoker"])
X_test['time'] = le3.transform(X_test["time"])
````

````python
X_test.head()
````

````python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Onehot encoder is applied with ColumnTransformer
````

````python
ct = ColumnTransformer(
    transformers=[
        ("OneHot", OneHotEncoder(drop="first"), [3])
    ],
    remainder="passthrough"
)

X_train = ct.fit_transform(X_train)
X_test = ct.transform(X_test)
````

````python
pd.DataFrame(X_train)
````

___
# Model  Training
___

````python
from sklearn.svm import SVR
````

````python
model =SVR()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
````

````python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
````

````python
print("R2 : ", r2_score(y_test, y_pred))
print("MSE : ", mean_absolute_error(y_test, y_pred))
````
