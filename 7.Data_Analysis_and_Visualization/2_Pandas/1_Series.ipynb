```python
import numpy as np
import pandas as pd
```

```python
# Creating A Series
nmbr  = [1,2,3,4,5]

series = pd.Series(nmbr) # in nmbr we Give Values , it assigned Indexes automatically
print(type(series))  # <class 'pandas.Series'>
print(series)

print()

series1 = pd.Series([34,45,67,78])
print(series1)
```

```python
# Custom Indexes 

marks = [56,78,45,89]
subjects = ["maths" , "Bio" , "Chem" , "DSA"]

marks_srs = pd.Series(marks, index = subjects) # instead of Zero based Indexing Now it will have subject values as index
print(marks_srs)
print()

rank = ["S", "S+", "A" , "SSS+"]
ninja = ["Diedara", "Itachi", "Neji" , "Hashirama"]
series3 = pd.Series(rank, index = ninja)
print(series3)
```

```python
# Series Using Dictionary
data = {
    "Naruto": "S+",
    "Sasuke": "S+",
    "Sakura": "A",
    "Kakashi": "SS"
}

srs = pd.Series(data) # Keys will be taken as Indexes
print(srs)
```

# Editing A series 

    series[idx] = new_value

- if that idex doesnt exist then a new item will be made

## Using Slicing 

    series[start, end, step] = [new values ]

    series[1:3] = [20,30]

## using Fancy Indexing

    series[[1,3]] = [24,765]

```python
marks = [56,78,45,89]
subjects = ["maths" , "Bio" , "Chem" , "DSA"]

grade = pd.Series(marks, index = subjects)
print(grade)
print()
print("Editing \n")
grade["maths"] = 91
print( grade["maths"]) # prints 91
grade["Bio"] = 12
print( grade["Bio"])
print(grade)  # changes 
print("\n index that doesnt exist")
grade["OOP"] = 42
grade["Web"] = 00

print(grade)  # OOP and Web Added
```

```python
# usig Slicing TO Edit 
print(vk)
vk[1:3] = [20,182]
print()
print(vk)

# using Fancy Indexing 
vk[[4,5]]  = [433,64]
print()
print(vk)
```
