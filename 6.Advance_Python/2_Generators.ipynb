# Generators #
Subclass of iterator itself. Simple Way to Create an Iterator.
* Generates Value on the Fly and Do not Store it Memory.
* use " yield " keyword

        A generator is a function that returns values one at a time using yield instead of returning everything at once.


| return                    | yield                  |
| ------------------------- | ---------------------- |
| Ends function             | Pauses function        |
| Runs once                 | Runs many times        |
| Returns value immediately | Produces values lazily |

    return does two things:

- 1️⃣ Sends a value back
- 2️⃣ Immediately stops the function

        Yield:

- Pause Function and Continue Later
- Give me one at a time


### Difference b/w Iterator and Generator ###

| Concept       | Explanation                                                                                                                                                              |
| ------------- | ------------------------------------------------------------------------------------------------------------------------------------------------------------------------ |
| **Iterator**  | An object that implements the **iterator protocol**: has `__iter__()` and `__next__()`. It returns values one by one. You can make **custom iterators** using classes.   |
| **Generator** | A **special type of iterator** created using a **function with `yield`** or a generator expression `( )`. Python automatically implements `__iter__()` and `__next__()`. |


```python
## Normal Function 

def Count_n(n):
    for i in range(1,n+1): 
        return i # 1st it will return 1 , but when it does ... Function Stops


n = Count_n(6)
print(n) # prints Only 1 not the whole LOOP from 1-6
         # Return Does Two Things:
            # Sends Value Back
            # Immedietly stops the function

#Count_n(6)
# Loop start
#  |
#  |
# i = 1
#  |
#  v
# return 1
#  |
#  v
# FUNCTION STOPS
```

```python
## The Above Function Using Yield 

def Count_1_n(n):
    for i in range(1,n+1):
        yield i  # Pause function and Continue Later
                 # instead of all the Values Give me One Just For Now

gt  = Count_1_n(6)
print(type(gt))  # <class 'generator'>

print(next(gt))
print(next(gt))
print(next(gt))
print(next(gt))
print(next(gt))
print(next(gt))


# yield 1 → pause  -->> python resume the Function from where its last Paused
# yield 2 → pause
# yield 3 → pause
# yield 4 → pause
# yield 5 → pause
# yield 6 → pause
# StopIteration

## using Loop 

print("using Loop")
for n in gt:  # it won't run because gt has printed all the values ( as it also has one value at a time (not whole thing))
    print(next(n))

gt2 = Count_1_n(8)
for n in gt2:  # The for loop calls next() automatically.
    print(n)
```

```python
## WE can Return Multiple Yeild Values 

def My_generator():
    yield 1
    yield 2
    yield 3

gen = My_generator()
print(next(gen))
print(next(gen))  # each next  Moves to next yield
print(next(gen))
```

```python
def Sequence():
    yield "start"

    for i in range(5):
        yield i

    yield "End"

gen2 = Sequence()
for i in gen2:
    print(i)
    
```

```python
def test():
    yield 1
    yield 2
    yield 3

g = test()

print(next(g)) # 1
print(next(g)) # 2

for x in g:
    print(x)  # 3 as each next moves to next yield.. above we move 2
```

# Practical Examples #


```python
# image A sys that Generates User IDs One by One and each time starts from prevois state .. not from zero
def gen_Id():
    id = 1
    while True:
        yield id
        id += 1

ids = gen_Id()
print(next(ids))
print(next(ids))
print(next(ids))
```

```python
# Image you have 20gb file .. we whouldnt want to load at it once in memory
# there we can use yield and loads only one line at a time then move to next

def read_large_file(file_path):
    with open(file_path, "r") as file:
        for line in file:
            yield line

file_path = "large_file.txt"
gen3 = read_large_file(file_path)

for line in gen3:
    print(line.strip())
```

```python
# Filtering Data ... Imagine A data Pipeline
# numbers → filter even → square them

def numbers():
    for i in range(10):
        yield i

# filter Even

def even(nums):
    for n in nums:
        if n%2 == 0:
            yield n

#square Them
def Sq(nums):
    for n in nums:
        yield n*n

## Implementation
data = numbers()          # each value processed one-by-one
evens = even(data)
squares = Sq(evens)

for s in squares:
    print(s)
```

# Using Iterator and Generator Togather #

Scenario: Processing Huge Sensor Data for IoT

Imagine a company has IoT sensors collecting temperature and humidity in thousands of factories.

Each factory has millions of readings per day.

You need to process this data for reporting and alerts.

We’ll see how generators and iterators can work together.

```python
## Using Iterator and Generators Togather
import random

def sensor_data(num_sensors):
    """Yields Reading from Sensors Continuesly"""

    for sensor_id in range(1, num_sensors+1):
        # generate some random readings
        yield {
            sensor_id: sensor_id,
            "temperature": random.randint(20, 40),
            "humidity": random.randint(30, 80)
        }


## Custom Iterator
class HighTempAlert:
    def __init__(self, readings, threshold):
        self.readings = readings
        self.threshold = threshold
        self.idx = 0
        self.filtered = list(readings)

    def __iter__(self):
        return self 
    
    def __next__(self):
        while self.idx < len(self.filtered):
            reading = self.filtered[self.idx]
            self.idx += 1

            if reading["temperature"] > self.threshold:
                return reading
            
        raise StopIteration



data_stream = sensor_data(5)
for reading in data_stream:
    print(reading)

# Generator produces data
data_stream = sensor_data(10)

# Iterator filters data
alerts = HighTempAlert(data_stream, threshold=30)

for alert in alerts:
    print(f"ALERT! Sensor {list(alert.keys())[0]} temp={alert['temperature']}")
```
