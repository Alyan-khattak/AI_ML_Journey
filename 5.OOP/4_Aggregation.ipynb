### Aggregation ###

* One Class Owns Another Class.
* Has-A relation -->> Customer has Address.
* weak relation, both can exist Independently.
* Pass objct from outside.

## Basic Syntax ###
    Creating object of one class
    Passing it into another class


```python
### Customer Has Address 

class Customer:
    def __init__(self,name, gender, Address): # name,gender is simple(simple text )but address is complex like it has city,street,code etc
        self.name = name
        self.gender = gender
        self.Address = Address

    ## printing Address Here
    def print_add(self):          # it cant access private attribute like city so accessing using getter
        print(self.Address.street, self.Address.get_city(), self.Address.pin)
        #access Address object attributes so self.Address.attribute

    def edit_profile(self, new_name, new_street, new_city, new_pin):
        self.name = new_name
        #now to edit address we do have method in below class so use that
        self.Address.edit_add(new_street, new_city, new_pin)


## class 2 
class Address:

    def __init__(self,street, city, pin):
        self.street = street
        self.__city = city #private ( aggregation can access private attribute {above class can't access city})
        self.pin = pin

    #getter to print city
    def get_city(self):
        return self.__city

    # to edit address
    def edit_add(self, new_street, new_city, new_pin):
        self.street = new_street
        self.__city = new_city
        self.pin = new_pin


#Address Object
a1 = Address("78","Islambad",3823)


#Ctomer Object
c1 = Customer("Alyan","male",a1) # pass Address class objct a1
print(c1.print_add())

# edit profile
c1.edit_profile("shayan","73","sharjah",3842)
print(c1.print_add())

        
```

```python
class Team:

    def __init__(self, name, players):
        self.name = name
        self.players = players   # list of Player objects players = [p1,p2,p3]

    def start_match(self):
        print("Team", self.name, "is starting match")

        for player in self.players:
            player.play()


class Player:
    def __init__(self, name):
        self.name = name
    
    def play(self):
        print(self.name, "is playing")


p1 = Player("Alyan")
p2 = Player("Adil")
p3 = Player("Haroon")

team1 = Team("SE", [p1, p2, p3])
team1.start_match()
```
